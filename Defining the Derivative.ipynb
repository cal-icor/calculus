<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">1. Secant Lines and Tangent Lines</h1>

In Calculus, we are interested in understanding how functions change. Two fundamental ideas that help us explore this area are **secant lines** and **tangent lines**.

Revisiting the notion of secant lines and tangent lines. A secant line is a line that intersects a curve at two points. It gives an average rate of change of the function between those two points. By picking a point $(a,f(a))$ and choosing a value of $x$ near $a$, the secant line will be the line drawn throung the points $(a, f(a))$ and $(x,f(x))$ (see picture below). The slope of this line is given by an equation in the form of a difference quotient:

$$m_{sec} = \frac{f(x)-f(a)}{x-a}$$

By replacing $x$ with $a+h$, where $h$ is a value close to $0$. We can rewrite the equation above to calculate the the slope of the line through two points $(a,f(a))$ and $(a+h, f(a+h))$:

$$m_{sec} = \frac{f(a+h)-f(a)}{a+h-a} = \frac{f(a+h)-f(a)}{h}$$


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    <h2>Slope of Secant Line </h2>

Let $f(x)$ be a function. The secant line between the points $(a,f(a))$ and $(x,f(x))$ on the graph of $f$ has a slope given by:

$$m_{sec} = \frac{f(x)-f(a)}{x-a}$$

If we let $x=a+h$, where $h \neq 0$, then:

$$m_{sec} = \frac{f(a+h)-f(a)}{h}$$

is a difference quotient with increment $h$
</div>

From the picture above, as the value of $x$ gets closer to $a$, the slope of the secant line becomes a better approximation of the function’s rate of change at $a$. In fact, the **secant lines** gradually become closer to the **tangent line** at $a$ which represents the limiting case of the secant slopes. Likewise, as $h$ approaches $0$, the secant lines get closer to the tangent line. The slope of the tangent line at $a$ gives the instantaneous rate of change of the function at that point (see picture below).

<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2> Example #1.1</h2>

Consider the function $f(x) = x^2$ and its tangent line at the point (1,1) are shown over increasingly smaller intervals around $x=1$. As the intervals become narrower, the graph of the function and its tangent line begin to overlap, meaning the tangent line provides a good estimate for the function's values when $x$ is near $1$. This shows that the graph of $f(x)$ looks almost linear very close to $x=1$.

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    <h2> Slope of Tangent Line </h2>
    
    
Consider the function $f(x)$ defined in an open interval containing $a$. The tangent line at $x=a$ has a slope given by the limit of the secant slope as $x \to a$:

$$m_{tan} = \displaystyle\lim_{x \to a} \frac{f(x)-f(a)}{x-a}$$

Or equivalently, using $h \to 0$:

$$m_{tan} = \displaystyle\lim_{h \to 0} \frac{f(a+h)-f(a)}{h}$$

</div>

<h2> Remark: </h2> 

The slope of the tangent line is defined using two different forms, just as the slope of a secant line is defined using two different formulas.  We use both definitions in this text.  As before, the setting will determine which definition is used.  We may now find equations of tangent lines using the formal definition of a tangent line to a function at a point.

<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2>Example #1.2: </h2>

Consider the same function above $f(x) = x^2$, but we pick the point $(2,4)$ for this example. By taking the limit as $x$ approaches $2$, the point $(x,f(x))$ in the definition is getting closer and closer to the point $(2,4)$.

<h2>Solution </h2>

Let's find the limit:

$$\displaystyle\lim_{x \to 2} \frac{f(a)-f(2)}{x-2} = \displaystyle\lim_{x \to 2} \frac{x^2-4}{x-2} = \displaystyle\lim_{x \to 2} \frac{(x+2)(x-2)}{x-2} = \displaystyle\lim_{x \to 2} (x+2) = 2 + 2 = 4$$

This means that the true and exact slope of the curve (or slope of tangent line) at $x=2$ is 4.

<h2>Python</h2>

Let's explore this example using Python. In the cell below, we create 2 arrays: one for values of $x$ and one for function's value at each value of $x$. Run the code below *(press shift and enter at the same time)* to see the closer $x$ is to $2$, the more accurate the slope is.

```python
#Load the libraries
import numpy as np

# Create an array of x values that approach a from below

a = 2  # Point of differentiation
x_values = np.linspace(1.5, 2, 25, endpoint=False)  # values of x starting at 1.5 to 2

# Define the function
def f(x):
    return x**2  # f(x) = x^2

# Compute slopes by using the definition 
def slope(x_values):
    return (f(x_values)-f(a))/(x_values-a)

y_values = [slope(x) for x in x_values]


print("Restults of limits:")
print("x \t Slopes")
for x, y in zip(x_values, y_values):
    print(f"{x:.3f}\t {y:.3f}")
```

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">2. The Derivative of a Function at a Point</h1>


The **derivative** of a function represents the slope of the function at a given point, denoted by $f'(x)$. From Section 1, we learned that the slope at a point on a curve corresponds to the slope of the tangent line at that point. Therefore, finding the derivative at a point is equivalent to finding the slope of the tangent line there. The process of finding a derivative is called **differentiation**.

From algebra we know a lot about slopes, but lets focus on a few of key properties of slopes:
- Lines are straight and have constant slope
- When the line has a positive slope, the function values are increasing and when the line has a negative slope, the function values are decreasing.
- If we know any two points on a line, we can calculate the slope, $m$, using this formula: $m=\displaystyle\frac{y_2-y_1}{x_2-x_1}$

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition of Derivative</h2>

Let $f(x)$ be a function defined is an open interval containing $a$. The derivative of the function $f(x)$ at $a$, denoted by $f'(x)$, is defined by:
    
$$f'(a) = \displaystyle\lim_{x \to a} \frac{f(x)-f(a)}{x-a}$$
    
provided this limit exists.
    
Alternatively, we may also define the derivative of $f(x)$ at $a$ as:
    
$$f'(a) = \displaystyle\lim_{h \to 0} \frac{f(a+h)-f(a)}{h}$$

</div>

```python
#Load the libraries
import numpy as np

# Create an array of x values that approach a from below

a = 2  # Point of differentiation
x_values = np.linspace(1.5, 2, 25, endpoint=False)  # values of x starting at 1.5 to 2

# Define the function
def f(x):
    return 4*(x**2) - x + 2  # f(x) = 4x^2 - x + 2

# Compute slopes by using the definition 
def slope(x_values):
    return (f(x_values)-f(a))/(x_values-a)

y_values = [slope(x) for x in x_values]


print("Restults of limits:")
print("x \t Slopes")
for x, y in zip(x_values, y_values):
    print(f"{x:.3f}\t {y:.3f}")
```

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">3. Equation of Tangent line</h1>

In Section 1, we learned what a tangent line is, and in Section 2, we explored how to use the derivative of a function $f(x)$ to find the slope of the tangent line at a point. Now, a natural question arises: How can we find the full equation of the tangent line at a specific point on the graph of a function?

Let see the definition below:

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Equation of Tangent line</h2>
    
   Let $f(x)$ be a differentable function and $a$ a point in its domain. The slope of the tangent line at $x=a$ is given by the derivative $f'(a)$. The point of tangency is $(a,f(a))$. The equation of the tangent line is derived using the point-slope form of a line:
    
$$y - f(a) = f'(a)(x-a)$$
    
Or equivalently,
    
$$y = f'(a)(x-a)+f(a)$$

    
    
</div>


<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2> Example #3</h2>

**Find the equation of tangent line to $f(x) = x^2$ at $x=3$.**

<h2> Solution </h2>

- Compute $f(3)$:

$$f(x)= 3^2 = 9$$

- Compute the derivative $f'(x)$:

$$f'(x) = 2x \implies f'(3) = 2(3) = 6$$

- Use the point-slope formula:

$$y-9 = 6(x-3) \implies y = 6x - 9$$

**Final Asnwer**: The equation of tangent line is $y = 6x - 9$

<h2>Python </h2>

Let's use Python and SymPy to compute the tangent line to $f(x) = x^2$ at $x=3$

```python
import sympy as sp

# Define the variable
x = sp.Symbol('x')

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the point of tangency
a = 3
fa = f(a) #f(3)

# Compute the derivative symbolically
f_expr = f(x)
f_prime = sp.diff(f_expr, x) #using .diff() to find derivative of function
slope = f_prime.subs(x, a) #substitute a to f(x) to find the slope f'(a) at x=a

# Write the tangent line: y = f'(a)(x - a) + f(a)
tangent_line = slope * (x - a) + fa

# Display results
print("Function:", f_expr)
print("Derivative:", f_prime)
print("Point of tangency: ({}, {})".format(a,fa))
print("Slope of tangent line:", slope)
print("Equation of tangent line:", tangent_line)
```

Let's explore how we can draw the plot of function $f(x)$ and its tangent line at $x=3$ by using matplotib and numpy from Python

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Symbolic setup
x = sp.Symbol('x')

# Define the function
def f(x):
    return x**2

# Point of tangency
a = 3
fa = f(a)

# Derivative (slope of tangent line)
f_expr = f(x)
f_prime = sp.diff(f_expr, x)
slope = f_prime.subs(x, a)

# Tangent line equation: y = slope * (x - a) + f(a)
def tangent_line(x_val):
    return slope * (x_val - a) + fa

# Generate x-values for plotting
x_vals = np.linspace(0, 6, 400)
f_vals = f(x_vals)
tan_vals = tangent_line(x_vals)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_vals, label='$f(x) = x^2$', color='blue')
plt.plot(x_vals, tan_vals, label='Tangent at $x=3$', color='red', linestyle='--')
plt.scatter(a, fa, color='black')  # Point of tangency
plt.text(a, fa + 1, f'({a}, {fa})', ha='center')

plt.title('Graph of $f(x) = x^2$ and its Tangent Line at $x = 3$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.show()
```

This interactive tool helps you visualize the concept of a tangent line to the curve $f(x)=x^2$.

Use the slider labeled “x-coordinate” to choose the point on the curve where you'd like to draw the tangent line. As you move the slider, the red dot shows the selected point, and the orange dashed line updates to show the tangent line at that point. This visually demonstrates how the slope of the tangent line changes depending on where you are along the curve.

This is a great way to connect geometric intuition with calculus concepts like derivatives and instantaneous rate of change.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
%matplotlib inline

# Define the function and its derivative
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# Plotting function with adjustable tangent point
def plot_tangent(x_val):
    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot the function
    x = np.linspace(-1, 3, 300)
    y = f(x)
    ax.plot(x, y, label=r"$f(x) = x^2$", color='darkblue', linewidth=2)

    # Plot the tangent line
    slope = f_prime(x_val)
    y_tangent = slope * (x - x_val) + f(x_val)
    ax.plot(x, y_tangent, '--', color='orange', linewidth=2, label='Tangent Line')

    # Mark the point of tangency
    ax.plot(x_val, f(x_val), 'ro', label=f'Tangent at x = {x_val:.2f}')

    # Labels, grid, and fixed axis limits
    ax.set_title("Tangent Line to $f(x) = x^2$", fontsize=14)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.set_xlim([-1, 3])  # Fix x-axis range
    ax.set_ylim([-1, 10])  # Fix y-axis range
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Create the slider for x-value
x_slider = FloatSlider(value=1.0, min=-1, max=3, step=0.1, description='x-coordinate:')
interact(plot_tangent, x_val=x_slider);
```

    Here is the step by step to find <b>the equation of tangent line</b> for function $f(x)$ at the point $x=a$:
    
- Compute $f(a)$: substitute $a$ to $x$ in function $f(x)$.
    
- Find the derivative $f'(a)$: Compute the derivative $f'(x)$, then substitute $x=a$ to find $f'(a)$
    
- Use the point-slope formula: $y - f(a) = f'(a)(x-a)$
