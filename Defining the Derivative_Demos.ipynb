## 1. Secant and tangent lines

- **Secant line**: through $(a, f(a))$ and $(a+h, f(a+h))$. Slope: $m_{\text{sec}} = \frac{f(a+h)-f(a)}{h}$.
- **Tangent line**: the line the secant approaches as $h \to 0$. Its slope is the **limit** of that quotient.

**Use the sliders below.** Move $h$ toward 0 and watch the secant line. Then answer the question in a **Markdown** cell.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
%matplotlib inline

def f(x):
    return x**2

def secant_tangent_demo(a=1, h=0.5):
    x = np.linspace(-0.5, 3, 200)
    plt.figure(figsize=(8, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x)=x^2$')
    fa = f(a)
    ah = a + h
    fah = f(ah)
    slope_sec = (fah - fa) / h if abs(h) > 1e-10 else 2*a
    x_sec = np.linspace(a - 0.5, ah + 0.5, 2)
    y_sec = fa + slope_sec * (x_sec - a)
    plt.plot(x_sec, y_sec, 'g--', lw=2, label='Secant')
    slope_tan = 2 * a
    x_tan = np.linspace(a - 0.8, a + 0.8, 2)
    y_tan = fa + slope_tan * (x_tan - a)
    plt.plot(x_tan, y_tan, 'r-', lw=2, label='Tangent at $x=a$')
    plt.plot(a, fa, 'ko', markersize=10)
    plt.plot(ah, fah, 'go', markersize=8)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend()
    plt.title(f'$a={a}$, $h={h:.3f}$; slope of secant = {slope_sec:.4f}')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

interact(secant_tangent_demo,
         a=FloatSlider(value=1, min=0.2, max=2, step=0.1, description='$a$'),
         h=FloatSlider(value=0.5, min=-0.99, max=0.99, step=0.05, description='$h$'));
```

## 2. Definition of the derivative

**Key idea:** The derivative of $f$ at $a$ is the limit of the difference quotient:

$$f'(a) = \lim_{h \to 0} \frac{f(a+h)-f(a)}{h}$$

Use the demo below. Move $h$ toward 0 and watch the numeric value. Then answer the question in a **Markdown** cell.

```python
from ipywidgets import FloatSlider, Dropdown, interact

def diff_quotient_demo(func_choice='x^2', a=1, h=0.2):
    if func_choice == 'x^2':
        def f(x): return x**2
        f_prime_a = 2*a
    elif func_choice == '4x^2 - x + 2':
        def f(x): return 4*x**2 - x + 2
        f_prime_a = 8*a - 1
    else:
        def f(x): return x**2
        f_prime_a = 2*a
    if abs(h) < 1e-12:
        h = 1e-12
    q = (f(a+h) - f(a)) / h
    print(f'Function: {func_choice},  a = {a},  h = {h}')
    print(f'Difference quotient (f(a+h)-f(a))/h = {q:.6f}')
    print(f'Limit f\'(a) = {f_prime_a}')

interact(diff_quotient_demo,
         func_choice=Dropdown(options=['x^2', '4x^2 - x + 2'], description='f(x)'),
         a=FloatSlider(value=1, min=0, max=3, step=0.5, description='a'),
         h=FloatSlider(value=0.2, min=0.001, max=0.5, step=0.01, description='h'));
```

## 3. Tangent line equation

**Key idea:** The tangent line at $x=a$ has slope $f'(a)$ and equation:
$$y - f(a) = f'(a)(x - a)$$

Use the demo below. Change $a$ and observe the tangent line. Then answer the question in a **Markdown** cell.

```python
def tangent_line_demo(func_choice='x^2', a=1):
    if func_choice == 'x^2':
        def f(x): return x**2
        slope = 2*a
        label = r'$f(x)=x^2$'
    else:
        def f(x): return 4*x**2 - x + 2
        slope = 8*a - 1
        label = r'$f(x)=4x^2-x+2$'
    x = np.linspace(-0.5, 2.5, 200)
    plt.figure(figsize=(8, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=label)
    fa = f(a)
    x_tan = np.linspace(a - 1, a + 1, 2)
    y_tan = fa + slope * (x_tan - a)
    plt.plot(x_tan, y_tan, 'r-', lw=2, label=f'Tangent: slope = {slope}')
    plt.plot(a, fa, 'ko', markersize=10)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend()
    plt.title(f'Tangent at $x={a}$: $y - {fa} = {slope}(x - {a})$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

from ipywidgets import Dropdown
interact(tangent_line_demo,
         func_choice=Dropdown(options=['x^2', '4x^2 - x + 2'], description='$f(x)$'),
         a=FloatSlider(value=1, min=0.2, max=2, step=0.1, description='$a$'));
```

## Summary

- **Secant slope:** $m_{\text{sec}} = \frac{f(a+h)-f(a)}{h}$. As $h \to 0$, the secant approaches the **tangent line**.
- **Derivative:** $f'(a) = \lim_{h \to 0} \frac{f(a+h)-f(a)}{h}$ (limit of the difference quotient).
- **Tangent line:** $y - f(a) = f'(a)(x - a)$.
