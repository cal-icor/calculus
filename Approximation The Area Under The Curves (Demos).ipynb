## 1. Sigma (summation) notation

**Key idea:** Sigma notation $\sum_{i=a}^{b} f(i)$ means add $f(a) + f(a+1) + \cdots + f(b)$. For example, $\sum_{i=1}^{n} i = 1 + 2 + \cdots + n = \frac{n(n+1)}{2}$.

**Use the demo below.** Change $n$ and see the sum and its value.

```python
from ipywidgets import interact, IntSlider
from IPython.display import display, Math

def sigma_demo(n=5):
    total = sum(i for i in range(1, n + 1))
    formula = n * (n + 1) // 2
    terms = ' + '.join(str(i) for i in range(1, n + 1))
    if n > 8:
        terms = '1 + 2 + \\ldots + ' + str(n)
    display(Math(r'\sum_{i=1}^{n} i = ' + terms + ' = ' + str(total)))
    display(Math(r'\frac{n(n+1)}{2} = \frac{%s(%s+1)}{2} = \frac{%s \cdot %s}{2} = %s' % (n, n, n, n+1, total)))
    print(f'For n = {n}: sum = {total}  (formula: n(n+1)/2 = {formula})')

interact(sigma_demo, n=IntSlider(value=5, min=1, max=20, step=1, description='n'));
```

## 2. Rectangles approximate area

**Key idea:** To approximate the area under $y = f(x)$ from $x=a$ to $x=b$, divide $[a,b]$ into $n$ subintervals of width $\Delta x = (b-a)/n$, build a rectangle on each subinterval (height = function value at a point in the subinterval), and add the areas.

**Use the demo below.** Change $n$, $a$, $b$, and the function. Watch how the rectangles approximate the area.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, Dropdown
from IPython.display import display, Math
%matplotlib inline

def rectangles_demo(func_choice='x^2', a=0, b=2, n=6):
    if func_choice == 'x^2':
        def f(x): return x**2
        label = r'$f(x)=x^2$'
    elif func_choice == 'sqrt(x)':
        def f(x): return np.sqrt(np.maximum(x, 0))
        label = r'$f(x)=\sqrt{x}$'
    else:
        def f(x): return x**2
        label = r'$f(x)=x^2$'
    
    a, b = min(a, b), max(a, b)
    n = max(1, int(n))
    dx = (b - a) / n
    x_edges = np.linspace(a, b, n + 1)
    
    # Left-endpoint heights
    x_left = x_edges[:-1]
    y_left = f(x_left)
    area_left = np.sum(y_left * dx)
    
    x_plot = np.linspace(a - 0.1 * (b - a), b + 0.1 * (b - a), 300)
    plt.figure(figsize=(9, 5))
    plt.plot(x_plot, f(x_plot), 'b-', lw=2, label=label)
    plt.axhline(0, color='k', lw=0.5)
    for i in range(n):
        xs = [x_edges[i], x_edges[i], x_edges[i+1], x_edges[i+1]]
        ys = [0, y_left[i], y_left[i], 0]
        plt.fill(xs, ys, alpha=0.4, edgecolor='green')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.title(f'Left-endpoint approximation: $n={n}$, $\\Delta x = {dx:.4f}$, total area $\\approx {area_left:.4f}$')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    display(Math(r'\text{Left-endpoint sum} \approx ' + f'{area_left:.6f}'))

interact(rectangles_demo,
         func_choice=Dropdown(options=['x^2', 'sqrt(x)'], value='x^2', description='f(x)'),
         a=FloatSlider(value=0, min=0, max=2, step=0.25, description='a'),
         b=FloatSlider(value=2, min=0.5, max=4, step=0.25, description='b'),
         n=IntSlider(value=6, min=1, max=30, step=1, description='n'));
```

## 3. Riemann sums (left, right, midpoint)

**Key idea:** We can choose the height of each rectangle in three common ways: **left** endpoint of the subinterval, **right** endpoint, or **midpoint**. Each gives a Riemann sum approximation; the midpoint rule often gives a better estimate for the same $n$.

**Use the demo below.** Switch between left, right, and midpoint. Compare the approximate areas.

```python
def riemann_demo(method='Left', a=0, b=2, n=8):
    def f(x): return x**2
    a, b = min(a, b), max(a, b)
    n = max(1, int(n))
    dx = (b - a) / n
    x_edges = np.linspace(a, b, n + 1)
    
    if method == 'Left':
        x_pts = x_edges[:-1]
        area = np.sum(f(x_pts) * dx)
        title_extra = 'left endpoints'
    elif method == 'Right':
        x_pts = x_edges[1:]
        area = np.sum(f(x_pts) * dx)
        title_extra = 'right endpoints'
    else:
        x_pts = (x_edges[:-1] + x_edges[1:]) / 2
        area = np.sum(f(x_pts) * dx)
        title_extra = 'midpoints'
    
    y_pts = f(x_pts)
    x_plot = np.linspace(a - 0.1 * (b - a), b + 0.1 * (b - a), 300)
    plt.figure(figsize=(9, 5))
    plt.plot(x_plot, f(x_plot), 'b-', lw=2, label=r'$f(x)=x^2$')
    plt.axhline(0, color='k', lw=0.5)
    for i in range(n):
        if method == 'Left':
            x0, x1 = x_edges[i], x_edges[i+1]
            h = f(x0)
        elif method == 'Right':
            x0, x1 = x_edges[i], x_edges[i+1]
            h = f(x1)
        else:
            x0, x1 = x_edges[i], x_edges[i+1]
            h = f((x0 + x1) / 2)
        xs = [x0, x0, x1, x1]
        ys = [0, h, h, 0]
        plt.fill(xs, ys, alpha=0.4, edgecolor='green')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.title(f'{method}-endpoint approximation: {title_extra}, area $\\approx {area:.4f}$')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    true_val = (b**3 - a**3) / 3
    print(f'{method} sum = {area:.6f}; true integral = {true_val:.6f}')

interact(riemann_demo,
         method=Dropdown(options=['Left', 'Right', 'Midpoint'], value='Left', description='Rule'),
         a=FloatSlider(value=0, min=0, max=2, step=0.25, description='a'),
         b=FloatSlider(value=2, min=0.5, max=4, step=0.25, description='b'),
         n=IntSlider(value=8, min=2, max=25, step=1, description='n'));
```

## Summary

- **Sigma notation:** $\sum_{i=a}^{b} f(i) = f(a) + f(a+1) + \cdots + f(b)$.
- **Approximating area:** Divide $[a,b]$ into $n$ subintervals of width $\Delta x = (b-a)/n$; approximate area $\approx \sum_{i=1}^{n} f(\text{sample}_i) \cdot \Delta x$.
- **Riemann sums:** Left ($f(x_{i-1})$), right ($f(x_i)$), or midpoint ($f(m_i)$) rules. As $n \to \infty$, all three approach the definite integral $\int_a^b f(x)\,dx$.
