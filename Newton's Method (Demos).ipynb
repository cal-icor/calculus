## 1. How Newton's Method works

**Key idea:** To find a root of $f(x) = 0$, start with an initial guess $x_0$. The tangent line at $(x_0, f(x_0))$ crosses the $x$-axis at a new point $x_1$, which is (usually) closer to the actual root. Repeat this process:

$$x_{n} = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})}$$

**Use the demo below.** Adjust the initial guess $x_0$ and number of iterations. Watch how the method converges to a root.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider, Dropdown
%matplotlib inline

def newtons_method_demo(func_choice='x^2 - 2', x0=1.0, num_iterations=3):
    # Define function and its derivative
    if func_choice == 'x^2 - 2':
        def f(x): return x**2 - 2
        def fp(x): return 2*x
        x_range = (-1, 3)
        y_range = (-3, 5)
    elif func_choice == 'x^3 - x - 2':
        def f(x): return x**3 - x - 2
        def fp(x): return 3*x**2 - 1
        x_range = (-2, 3)
        y_range = (-5, 10)
    elif func_choice == 'cos(x) - x':
        def f(x): return np.cos(x) - x
        def fp(x): return -np.sin(x) - 1
        x_range = (-1, 2)
        y_range = (-2, 2)
    else:
        def f(x): return x**2 - 2
        def fp(x): return 2*x
        x_range = (-1, 3)
        y_range = (-3, 5)
    
    # Perform Newton's method iterations
    iterations = [x0]
    for i in range(num_iterations):
        if abs(fp(iterations[-1])) < 1e-10:
            print(f"Stopped at iteration {i}: derivative is zero at x = {iterations[-1]:.6f}")
            break
        x_new = iterations[-1] - f(iterations[-1]) / fp(iterations[-1])
        iterations.append(x_new)
    
    # Plot
    x = np.linspace(x_range[0], x_range[1], 300)
    plt.figure(figsize=(10, 6))
    plt.plot(x, f(x), 'b-', lw=2, label=f'$f(x) = {func_choice}$')
    plt.axhline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
    plt.axvline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
    
    # Plot each iteration
    colors = ['red', 'orange', 'purple', 'brown', 'pink']
    for i in range(len(iterations) - 1):
        x_curr = iterations[i]
        x_next = iterations[i+1]
        f_curr = f(x_curr)
        fp_curr = fp(x_curr)
        
        # Draw tangent line from (x_curr, f_curr) to (x_next, 0)
        x_tan = np.array([x_curr, x_next])
        y_tan = np.array([f_curr, 0])
        color = colors[i % len(colors)]
        plt.plot(x_tan, y_tan, color=color, linestyle='--', lw=1.5, 
                 label=f'Tangent at $x_{i}={x_curr:.4f}$')
        plt.plot(x_curr, f_curr, 'o', color=color, markersize=8)
        plt.plot(x_next, 0, 's', color=color, markersize=8)
    
    # Mark final point
    plt.plot(iterations[-1], f(iterations[-1]), 'go', markersize=10, 
             label=f'Final: $x_{len(iterations)-1}={iterations[-1]:.6f}$')
    
    plt.xlabel('$x$', fontsize=12)
    plt.ylabel('$y$', fontsize=12)
    plt.ylim(y_range)
    plt.legend(loc='best', fontsize=9)
    plt.title(f"Newton's Method: {len(iterations)-1} iterations", fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Print iteration values
    print(f"\nIterations:")
    for i, x_val in enumerate(iterations):
        print(f"  x_{i} = {x_val:.8f},  f(x_{i}) = {f(x_val):.8f}")

interact(newtons_method_demo,
         func_choice=Dropdown(options=['x^2 - 2', 'x^3 - x - 2', 'cos(x) - x'], 
                              value='x^2 - 2', description='f(x)'),
         x0=FloatSlider(value=1.0, min=-2, max=3, step=0.1, description='x_0'),
         num_iterations=IntSlider(value=3, min=1, max=5, step=1, description='Iterations'));
```

## 2. When Newton's Method fails

**Key idea:** Newton's Method doesn't always succeed. Three common failure modes:

1. **Zero derivative**: If $f'(x_n) = 0$, we divide by zero and the method breaks.
2. **Wrong root**: The method converges to a different root than intended.
3. **Cycling**: The iterations alternate between values without approaching any root.

**Use the demo below.** Select a failure mode and observe what goes wrong.

```python
def newtons_failure_demo(failure_mode='Zero derivative'):
    plt.figure(figsize=(10, 6))
    
    if failure_mode == 'Zero derivative':
        # f(x) = x^3 - 3x^2 + 2, f'(x) = 3x^2 - 6x
        def f(x): return x**3 - 3*x**2 + 2
        def fp(x): return 3*x**2 - 6*x
        x0 = 0.0  # f'(0) = 0
        x_range = (-0.5, 3.5)
        
        x = np.linspace(x_range[0], x_range[1], 300)
        plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x) = x^3 - 3x^2 + 2$')
        plt.axhline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
        plt.axvline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
        
        plt.plot(x0, f(x0), 'ro', markersize=12, label=f'$x_0={x0}$: f\'(x_0)=0')
        
        # Draw horizontal tangent line
        x_tan = np.linspace(x0 - 1, x0 + 1, 2)
        y_tan = np.array([f(x0), f(x0)])
        plt.plot(x_tan, y_tan, 'r--', lw=2, label='Tangent (horizontal)')
        
        plt.xlabel('$x$', fontsize=12)
        plt.ylabel('$y$', fontsize=12)
        plt.legend(fontsize=10)
        plt.title("Failure: Zero Derivative\nTangent line is horizontal, never crosses x-axis", fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.ylim(-2, 4)
        
        print(f"At x_0 = {x0}:")
        print(f"  f(x_0) = {f(x0):.4f}")
        print(f"  f'(x_0) = {fp(x0):.4f}")
        print(f"\nNext iteration: x_1 = x_0 - f(x_0)/f'(x_0) = {x0} - {f(x0)}/{fp(x0)} = undefined (division by zero)")
        
    elif failure_mode == 'Wrong root':
        # f(x) = x^3 - 6x^2 + 11x - 6 = (x-1)(x-2)(x-3)
        # Roots at x = 1, 2, 3
        def f(x): return x**3 - 6*x**2 + 11*x - 6
        def fp(x): return 3*x**2 - 12*x + 11
        x0 = 0.5  # Closer to root at x=1 than x=3
        x_range = (-0.5, 4)
        
        # Perform 4 iterations
        iterations = [x0]
        for i in range(4):
            x_new = iterations[-1] - f(iterations[-1]) / fp(iterations[-1])
            iterations.append(x_new)
        
        x = np.linspace(x_range[0], x_range[1], 300)
        plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x) = x^3 - 6x^2 + 11x - 6$')
        plt.axhline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
        
        # Mark the three roots
        plt.plot([1, 2, 3], [0, 0, 0], 'gs', markersize=10, label='Roots: x=1, 2, 3')
        
        # Plot iterations
        colors = ['red', 'orange', 'purple', 'brown']
        for i in range(len(iterations) - 1):
            x_curr = iterations[i]
            x_next = iterations[i+1]
            f_curr = f(x_curr)
            
            x_tan = np.array([x_curr, x_next])
            y_tan = np.array([f_curr, 0])
            color = colors[i % len(colors)]
            plt.plot(x_tan, y_tan, color=color, linestyle='--', lw=1.5)
            plt.plot(x_curr, f_curr, 'o', color=color, markersize=8)
        
        plt.plot(iterations[0], f(iterations[0]), 'ro', markersize=10, label=f'$x_0={x0}$')
        plt.plot(iterations[-1], f(iterations[-1]), 'mo', markersize=10, 
                 label=f'$x_4={iterations[-1]:.4f}$ (converged to x=1, not x=3)')
        
        plt.xlabel('$x$', fontsize=12)
        plt.ylabel('$y$', fontsize=12)
        plt.legend(fontsize=9, loc='upper right')
        plt.title("Failure: Converges to Wrong Root\nStarting at x_0=0.5 finds x=1 instead of x=3", fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.ylim(-2, 3)
        
        print(f"Wanted root at x = 3, but started at x_0 = {x0}")
        print(f"\nIterations:")
        for i, x_val in enumerate(iterations):
            print(f"  x_{i} = {x_val:.6f}")
        print(f"\nConverged to x ≈ 1 (wrong root!)")
        
    else:  # Cycling
        # f(x) = x^3 - 2x + 2, f'(x) = 3x^2 - 2
        def f(x): return x**3 - 2*x + 2
        def fp(x): return 3*x**2 - 2
        x0 = 0.0
        x_range = (-1.5, 2)
        
        # Show cycling: x0=0 -> x1=1 -> x2=0 -> ...
        iterations = [x0]
        for i in range(4):
            x_new = iterations[-1] - f(iterations[-1]) / fp(iterations[-1])
            iterations.append(x_new)
        
        x = np.linspace(x_range[0], x_range[1], 300)
        plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x) = x^3 - 2x + 2$')
        plt.axhline(0, color='k', linestyle='--', lw=0.5, alpha=0.5)
        
        # Plot iterations
        colors = ['red', 'orange', 'red', 'orange']
        for i in range(len(iterations) - 1):
            x_curr = iterations[i]
            x_next = iterations[i+1]
            f_curr = f(x_curr)
            
            x_tan = np.array([x_curr, x_next])
            y_tan = np.array([f_curr, 0])
            color = colors[i % len(colors)]
            plt.plot(x_tan, y_tan, color=color, linestyle='--', lw=1.5, alpha=0.7)
            plt.plot(x_curr, f_curr, 'o', color=color, markersize=8)
        
        plt.plot(0, f(0), 'ro', markersize=12, label='x=0')
        plt.plot(1, f(1), 'o', color='orange', markersize=12, label='x=1')
        
        plt.xlabel('$x$', fontsize=12)
        plt.ylabel('$y$', fontsize=12)
        plt.legend(fontsize=10)
        plt.title("Failure: Cycling Between Values\nIterations alternate: 0 → 1 → 0 → 1 → ...", fontsize=14)
        plt.grid(True, alpha=0.3)
        plt.ylim(-1, 4)
        
        print(f"Starting at x_0 = {x0}")
        print(f"\nIterations:")
        for i, x_val in enumerate(iterations):
            print(f"  x_{i} = {x_val:.6f}")
        print(f"\nThe method cycles between x=0 and x=1, never finding the actual root!")
    
    plt.tight_layout()
    plt.show()

interact(newtons_failure_demo,
         failure_mode=Dropdown(options=['Zero derivative', 'Wrong root', 'Cycling'], 
                               value='Zero derivative', description='Failure type'));
```

## Summary

- **Newton's Method formula:** $x_{n} = x_{n-1} - \frac{f(x_{n-1})}{f'(x_{n-1})}$ — iteratively use tangent lines to approximate roots.
- **When it works:** With a good initial guess $x_0$ near a root, the method converges quickly.
- **When it fails:**
  - **Zero derivative** ($f'(x_n) = 0$): Division by zero, tangent is horizontal.
  - **Wrong root**: Poor initial guess leads to a different root.
  - **Cycling**: Iterations alternate without converging.
