## 1. Mean Value Theorem for Integrals

**Key idea:** If $f$ is continuous on $[a,b]$, then there is at least one $c$ in $[a,b]$ such that $f(c)(b-a) = \int_a^b f(x)\,dx$. So the **average value** $f_{\text{avg}} = \frac{1}{b-a}\int_a^b f(x)\,dx$ is equal to $f(c)$ for some $c$.

**Use the demo below.** Change $a$ and $b$; see the area under the curve and the rectangle of height $f(c)$ with the same area.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.integrate import quad
%matplotlib inline

def f(x):
    return x**2

def mvt_integral_demo(a=1, b=3):
    a, b = min(a, b), max(a, b)
    area, _ = quad(f, a, b)
    f_avg = area / (b - a)
    # f(c) = f_avg => c^2 = f_avg => c = sqrt(f_avg) in [a,b]
    c = np.sqrt(f_avg)
    if c < a or c > b:
        c = (a + b) / 2  # fallback
    
    x = np.linspace(0, 4, 300)
    x_fill = np.linspace(a, b, 200)
    
    plt.figure(figsize=(9, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x)=x^2$')
    plt.fill_between(x_fill, f(x_fill), alpha=0.3, label='Area under curve')
    plt.axhline(f_avg, color='green', linestyle='--', alpha=0.8, label=f'$f_{{avg}} = f(c) = {f_avg:.4f}$')
    plt.axvline(a, color='gray', linestyle=':', alpha=0.7)
    plt.axvline(b, color='gray', linestyle=':', alpha=0.7)
    plt.bar([(a+b)/2], [f_avg], width=(b-a), bottom=0, alpha=0.2, color='green', edgecolor='green', linewidth=2, label='Rectangle area = area under curve')
    plt.plot(c, f(c), 'ro', markersize=12, label=f'$c = {c:.4f}$, $f(c) = f_{{avg}}$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend(loc='upper right', fontsize=9)
    plt.title(f'MVT for Integrals: $\\int_{{{a}}}^{{{b}}} f = {area:.4f} = f(c)(b-a)$, $c \\approx {c:.3f}$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f'Area = {area:.4f}, f_avg = {area:.4f}/({b}-{a}) = {f_avg:.4f}, c = {c:.4f}')

interact(mvt_integral_demo,
         a=FloatSlider(value=1, min=0.2, max=2.5, step=0.25, description='a'),
         b=FloatSlider(value=3, min=1.5, max=4, step=0.25, description='b'));
```

## 2. FTC Part 1: Derivative of the area function

**Key idea:** If $F(x) = \int_a^x f(t)\,dt$, then $F'(x) = f(x)$. So the derivative of the "area so far" function is the original function.

**Use the demo below.** Move the slider for $x$; see the shaded area $F(x)$ and that $F'(x) = f(x)$.

```python
from IPython.display import display, Math

def ftc1_demo(a=0, x_val=1.5):
    def f(t): return t**2
    def F(x): return (x**3 - a**3) / 3  # antiderivative of t^2
    a, x_val = min(a, x_val), max(a, x_val)
    area, _ = quad(f, a, x_val)
    
    t = np.linspace(0, 3, 300)
    t_fill = np.linspace(a, x_val, 200)
    
    plt.figure(figsize=(9, 5))
    plt.plot(t, f(t), 'b-', lw=2, label=r'$f(t) = t^2$')
    plt.fill_between(t_fill, f(t_fill), alpha=0.4, label=f'$F(x) = \\int_{{{a}}}^{{{x_val}}} f(t)\\,dt = {area:.4f}$')
    plt.axvline(a, color='gray', linestyle='--', alpha=0.7)
    plt.axvline(x_val, color='red', linestyle='-', alpha=0.8)
    plt.plot(x_val, f(x_val), 'ro', markersize=12, label=f'$f(x) = f({x_val}) = {f(x_val):.4f} = F\'(x)$')
    plt.xlabel('$t$ (or $x$)')
    plt.ylabel('$y$')
    plt.legend(loc='upper right', fontsize=9)
    plt.title(f'FTC Part 1: $F(x) = \\int_a^x f(t)\\,dt$ implies $F\'(x) = f(x)$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    display(Math(r'F(x) = \int_{%s}^{%s} t^2 \, dt = %s' % (a, x_val, "%.4f" % area)))
    display(Math(r"F'(x) = f(x) = %s" % ("%.4f" % f(x_val))))

interact(ftc1_demo,
         a=FloatSlider(value=0, min=0, max=1.5, step=0.25, description='a'),
         x_val=FloatSlider(value=1.5, min=0.5, max=2.5, step=0.1, description='x'));
```

## 3. FTC Part 2: Evaluating definite integrals

**Key idea:** If $F$ is any antiderivative of $f$ (i.e. $F' = f$), then $\int_a^b f(x)\,dx = F(b) - F(a)$. So we can compute definite integrals by finding an antiderivative and evaluating at the endpoints.

**Use the demo below.** Change $a$ and $b$; see $F(b)-F(a)$ and the integral value.

```python
def ftc2_demo(a=1, b=2):
    def f(x): return x**2
    def F(x): return x**3 / 3  # antiderivative
    a, b = min(a, b), max(a, b)
    integral_val, _ = quad(f, a, b)
    Fb_minus_Fa = F(b) - F(a)
    
    x = np.linspace(0, 3, 300)
    x_fill = np.linspace(a, b, 200)
    plt.figure(figsize=(9, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x)=x^2$')
    plt.fill_between(x_fill, f(x_fill), alpha=0.4, label=f'$\\int_{{{a}}}^{{{b}}} f = F(b)-F(a)$')
    plt.axvline(a, color='gray', linestyle='--', alpha=0.7)
    plt.axvline(b, color='gray', linestyle='--', alpha=0.7)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend(loc='upper right', fontsize=9)
    plt.title(f'FTC Part 2: $\\int_{{{a}}}^{{{b}}} x^2\\,dx = F(b)-F(a) = {Fb_minus_Fa:.4f}$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f'F(x) = x^3/3.  F(b)-F(a) = {F(b):.4f} - {F(a):.4f} = {Fb_minus_Fa:.4f};  integral = {integral_val:.4f}')

interact(ftc2_demo,
         a=FloatSlider(value=1, min=0, max=2, step=0.25, description='a'),
         b=FloatSlider(value=2, min=0.5, max=3, step=0.25, description='b'));
```

## 4. Applications (brief)

If $v(t)$ is velocity, then the **distance traveled** from time $t=a$ to $t=b$ is $\int_a^b v(t)\,dt$. So integration recovers position (or displacement) from velocity. Use the slider to change the time interval and see the distance.

```python
def velocity_distance_demo(a=0, b=3):
    # v(t) = 2t (simple velocity); distance = int v = t^2, so from a to b: b^2 - a^2
    def v(t): return 2 * t
    a, b = min(a, b), max(a, b)
    distance, _ = quad(v, a, b)
    
    t = np.linspace(0, 4, 300)
    t_fill = np.linspace(a, b, 200)
    plt.figure(figsize=(8, 4))
    plt.plot(t, v(t), 'b-', lw=2, label=r'$v(t) = 2t$')
    plt.fill_between(t_fill, v(t_fill), alpha=0.4, label=f'Distance = $\\int_{{{a}}}^{{{b}}} v = {distance:.2f}$')
    plt.axvline(a, color='gray', linestyle='--', alpha=0.7)
    plt.axvline(b, color='gray', linestyle='--', alpha=0.7)
    plt.xlabel('$t$ (time)')
    plt.ylabel('$v(t)$ (velocity)')
    plt.legend()
    plt.title('Distance traveled = area under velocity curve')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f'Distance from t={a} to t={b}: {distance:.2f}')

interact(velocity_distance_demo,
         a=FloatSlider(value=0, min=0, max=2, step=0.5, description='a'),
         b=FloatSlider(value=3, min=1, max=4, step=0.5, description='b'));
```

## Summary

- **MVT for Integrals:** There exists $c \in [a,b]$ with $f(c)(b-a) = \int_a^b f(x)\,dx$; equivalently $f(c) = f_{\text{avg}}$.
- **FTC Part 1:** $F(x) = \int_a^x f(t)\,dt$ implies $F'(x) = f(x)$.
- **FTC Part 2:** $\int_a^b f(x)\,dx = F(b) - F(a)$ for any antiderivative $F$ of $f$.
- **Application:** Distance traveled = $\int_a^b v(t)\,dt$.
