<h1>Introduction</h1>

If our function variables have meaning, we can use the area under the graphs of these curves to solve real life application problems. For examaple, if a function $v(t)$ represents the velocity of an object from the time $a$ to $b$, then the **area under the curve** of $v(t)$ gives the total distance traveled during that time interval.  
We could imagine this by considering a velocity function with constant velocity of 65 mph, over an interval of 3 hours (see graph below).  If we calculate the area between  $v(t)=65$ and the x-axis, we would get: base X height = 3 hours X 65 miles/hour = 195 miles.  The area between $v(t)$ and hte x-axis represents the distance traveled over 3 hours.

Now let's consider another velocity function, shown below:

If we want to find the disance traveled over the interval [a,b], how can we calculate the area under the velocity function above?  This curve isn't a perfect shape like a rectangle, triangle or other shape that we know the area formula for. This is where approximating the area using rectangles comes in. The idea is to break the interval $a$ to $b$ into smaller subintervals, estimate the height of the curve on a point within each subinterval, calculate the area of the rectangle formed on each subinterval, and add them up to approximate the full area.

Before we do this, we need to understand **summation notation**, using the sigma symbol $\sum$. This allows us to express how we add up all the individual rectangle areas efficiently.

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">1. Sigma (summation) notation</h1>

<h2>Definition</h2>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #001f4d;">
  Sigma notation ( $\sum$ ) is a mathematical way to represent the sum of a sequence of terms. It is commonly used to express the sum of integers, powers of integers, and other series. The general form is:
<div style="text-align:center; font-size:20px;">$$\sum_{i=a}^{b} f(i)$$</div>
where:
<ul>
    <li>$i$ is the index of summation.</li> 
    <li>$a$ is the starting value. </li>
    <li>$b$ is the ending value.  </li>
    <li>$f(i)$ is the function to be summed.  </li>
<ul>

</div>

For the expression above, the sum will start with $i=a$, then add the function value when $i=a+1$, then $i=a+2$, and keep adding in function values until $i=b$.  This assumes that $i$ takes on integer values only.

<div style="text-align:center; font-size:20px;">$$\sum_{i=a}^{b} f(i) = f(a)+f(a+1)+f(a+2)+...+f(b-1)+f(b)$$</div>

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">2. Approximating the Area Under the Curve</h1>

Now we want to use the sum of the areas of rectangles to calculate the approximate area between the graph of a function and the x-axis.



<h3>Question: How can we approximate the area $A$ bounded by $f(x)$ above and the *x-axis* below, on the interval from $x=a$ to $x=b$? (See graph below)</h3>

<h3>Answer:</h3>

We can divide that area into many small shapes, each of which has a known area formula. Rectangles are a good choice, since we can easily calculate the area of a rectangle: Area = width x height.

If we want to create "$n$" rectangles, we define the width of the rectangles as $\Delta x$, where we take the length of the entire interval ($b-a$) and divide it into $n$ pieces: $\Delta x = \frac{b-a}{n}$.

Then, we can create equally spaced points $x_0, x_1, x_2, ..., x_n$ with $x_0=a, x_n=b$ and $x_i = x_0 + i \Delta x$ for $i = 1, 2, 3, ...,n.$  These values of x make up the endpoints of $n$ subintervals: $[x_0, x_1],   [x_1, x_2],   [x_2, x_3],  ...,   [x_n-1, x_n]$. 

Next we need to define the height of each rectangle.  We do this by calculating the function value an x-value within each subinterval. Our Calculus 1 textbook shows 3 ways to choose these x-values, though there are other ways.  We can use: **left-endpoint, right-endpoint, and midpoint approximation rules**.  As you might guess, for "left-endpoints" we use the smallest x-value in each subinterval (or the left most endpoint) to input into our function to define the height of each rectangle.  For "right-endpoints", we would use the largest x-value in each subinterval (or the right most endpoint), and for the "midpoint rule", we would use the x-value in the middle of each subinterval (you can average the end points to find the midpoint of the subinterval).

Using sigma notation, we can write out these Rules in an efficient way.


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    
<h2>Rule: Left-Endpoint Approximation</h2>
    
    
  We divide the $[a, b]$ into $n$ subinterval $[x_{i-1}, x_i]$ (with $i = 1,2,3,...,n$), then construct rectangles with width $\Delta x$ and height equal to $f(x_{i-1})$, which is the function value at the left endpoint of the subinterval. This makes the area of each rectangle $f(x_{i-1})\Delta x$. By adding the areas of all the rectangles we create, we get an approximate value for the area between f(x) and the x-axis on the interval $[a,b]$, or the area labeled $A$ below. We use the notation $L_n$ and sigma (summation) notation that we learned above to create the left-endpoint approximation of $A$ using $n$ subintervals.
<div style="text-align:center; font-size:20px;">
$$A \approx L_n = f(x_0)\Delta x + f(x_1)\Delta x+ ... + f(x_{n-1})\Delta x = \sum_{i=1}^{n} f(x_{i-1})\Delta x$$
</div>
</div>


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    
<h2>Rule: Right-Endpoint Approximation</h2>
    
    
  In the Right-endpoint approximation rule, we still use the same the witdh $\Delta x$ for each rectangle, by now we use the function value $f(x_i)$ at the right endpoint of the subinterval as the height of the rectangle. Therefore, the area of each rectangle is $f(x_i)\Delta x$ and the approximation for $A$ is:

<div style="text-align:center; font-size:20px;">
$$A \approx R_n = f(x_1)\Delta x + f(x_2)\Delta x+ ... + f(x_n)\Delta x = \sum_{i=1}^{n} f(x_i)\Delta x$$
</div>
</div>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    
<h2>Rule: Midpoint Approximation</h2>
    
    
  In this case, we use the function value $f(m_i)$ as the height of the rectangle, where $m_i$ is the the midpoint of the i-th subinterval $[x_i,x_{i+1}]$. 


Therefore, the area of each rectangle is $f(m_i)\Delta x$ and the approximation for $A$ is:
<div style=" font-size:20px;">
$$A \approx M_n = f(m_1)\Delta x + f(m_2)\Delta x+ ... + f(m_n)\Delta x = \sum_{i=1}^{n} f(m_i)\Delta x$$
</div>
</div>

<div style="border: 3px solid; border-image: linear-gradient(to right, #001f4d, #7f00ff, ) 1; padding: 15px; border-radius: 12px; font-family:'Segoe UI', sans-serif;">
<h2>Strategy: How to approximate the area between the x-axis and a given funciton $f(x)$</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>
    
**Step 1: Find the Width of Each Rectangle**

Divide the interval $[a,b]$ into n equal subintervals with width: $\Delta x = \frac{b-a}{n}$.  We will use this as the width of each rectangle.

**Step 2: Calculate the Heights**

<li>Left-hand endpoints: $x_i = a + (i - 1) \Delta x$</li>
<li>Right-hand endpoints: $x_i = a + i\Delta x$</li>
<li>Midpoints: $x_i = m_i$, where $m_i$ is the midpoint of the i-th subinterval $[x_i,x_{i+1}]$</li>

For whichever method you choose, the heights of each rectangle can be found by inputting each x-value into the function: $f(x_i)$

**Step 3: Compute and Sum the Rectangle Areas:**

We can calculate the approximate area between the curve and the x-axis by adding up the area of each rectangle formed by each subinterval.  To find the area of each rectangle, we multiply the width by the height. Depending on which approximation rule you chose, we use the following sums to approximate the area desired: 

The left-endpoint: 
<div style=" font-size:20px;">
$$A \approx L_n = \sum_{i=1}^{n} f(x_{i-1})\Delta x$$
</div>

The right-endpoint:
<div style=" font-size:20px;">
$$A \approx R_n = \sum_{i=1}^{n} f(x_i)\Delta x$$
</div>

The mid-endpoint:
<div style=" font-size:20px;">
$$A \approx M_n = \sum_{i=1}^{n} f(m_i)\Delta x$$
</div>


where:
<li>$n$: number of rectangles</li>
<li>$\Delta x = \frac{b-a}{n}$: the width of each rectangle</li>
<li>$f(x_i), f(x),$ and $f(m_i)$ : height of the rectangle</li>
    
</div>

**Approximate the area under the curve of function $f(x) = x^2$ in $[1,3]$ by dividing area into 4 rectangles and using the left points and right points.**

<h2> Solution </h2>

**The left-endpoint approximation rule:**

Step 1: 
Interval $[1,3]$, divide in to $n=4$ subintervals.
So $\Delta x = \frac{3-2}{4} = 0.5$

Step 2:
Using the left points: $x_0 = 1, x_1 = 1.5, x_2 = 2, x_3 = 2.5$

Step 3:
Approximate the area by adding up the rectangle areas:
$$A \approx L_4 = \sum_{i=1}^{4} f(x_{i-1})\Delta x = f(x_0)\Delta x + f(x_1)\Delta x + f(x_2)\Delta x + f(x_3)\Delta x$$
$$= (1)^2(0.5) + (1.5)^2(0.5) + (2)^2(0.5) + (2.5)^2(0.5) = 6.75$$

**The right-endpoint approximation rule:**

Step 1: 
Interval $[1,3]$, divide in to $n=4$ subintervals.
So $\Delta x = \frac{3-2}{4} = 0.5$

Step 2:
Using the left points: $x_1 = 1.5, x_2 = 2, x_3 = 2.5, x_4 = 3$

Step 3:
Approximate the area by adding up the rectangle areas:
$$A \approx R_4 = \sum_{i=1}^{4} f(x_i)\Delta x = f(x_1)\Delta x + f(x_2)\Delta x + f(x_3)\Delta x + f(x_4)\Delta x$$
$$= (1.5)^2(0.5) + (2)^2(0.5) + (2.5)^2(0.5) + (3)^2(0.5) = 10.75$$

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">3. Introduce the Riemann Sums</h1>


We have learned how to use rectangles to approximate the area $A$ between a curve and the x-axis by finding equal widths over an interval and heights defined by function values using righthand endpoints, lefthand endpoints, or midpoints of each subinterval $[x_{i-1},x_i]$. In fact, we can choose ANY point $x_i^*$ in each subinterval and use $f(x_i^*)$ as the height of each rectangle. This gives us an approximation of the area $A$ as:
<div style=" font-size:20px;">
$$A \approx \sum_{i=1}^{n} f(x_i^*)\Delta x$$
</div>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Riemann Sums</h2>
    
Let $f(x)$ be defined on a closed interval $[a,b]$ and let's assume we can create a regular partition of $[a,b]$, where $\Delta x$ is the width of each subinterval $[x_{i-1}, x_i]$ and for each $i$, let $x_i^*$ be any point in $[x_{i-1}, x_i]$. A **Riemann sum** is defined for $f(x)$ as:
<div style=" font-size:20px;">
$$\sum_{i=1}^{n} f(x_i^*)\Delta x$$
</div>

</div>

This sum gives us an appoximation for the area between the graph of f(x) and the x-axis.  As we mentioned before, the more rectangles we use, the more accurate this approximation will be. This is also true for Riemann sums; it will give better approximations for larger values of $n$. Therefore, we can approximate the area under a curve by using Riemann sums and if we take the limit of this sum as n approaches $oo$, creating infinitely many rectangles, this limit calculates the true area between f(x) and the x-axis over the interval.


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Area Under the Curve</h2>
    
Let $f(x)$ be a continuous, nonnegative function on a interval $[a,b]$, and let $\sum_{i=1}^{n} f(x_i^*)\Delta x$ be a Riemann sum for $f(x)$. Then, the **area under the curve** $y=f(x)$ on $[a,b]$ is given by
$$
A = \lim_{n \to \infty} \sum_{i=1}^{n} f(x_i^*) \Delta x
$$

</div>

True area found by definite integration: 
$$\int_0^1 x^2 \, dx = \left[ \frac{x^3}{3} \right]_0^1 = \frac{1}{3}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2

# Define the exact area using integration (for comparison)
exact_area = 1/3

# Define a function to compute left Riemann sum
def left_riemann(f, a, b, n):
    dx = (b - a) / n
    x_left = np.linspace(a, b - dx, n)
    area = np.sum(f(x_left) * dx)
    return area

# Define different values of n
n_values = [5, 10, 100, 1000]
areas = []

for n in n_values:
    approx_area = left_riemann(f, 0, 1, n)
    areas.append(approx_area)
    print(f"n = {n}, Approximate area = {approx_area:.6f}, Error = {abs(approx_area - exact_area):.6f}")

# Optional: Plotting convergence
plt.figure(figsize=(8, 5))
plt.plot(n_values, areas, 'o-', label='Approximate Area')
plt.axhline(y=exact_area, color='r', linestyle='--', label='Exact Area (1/3)')
plt.xlabel("Number of Rectangles (n)")
plt.ylabel("Approximate Area")
plt.title("Convergence of Left Riemann Sum for f(x) = x² on [0,1]")
plt.legend()
plt.grid(True)
plt.show()
```

<h2> Remark </h2>

The graph clearly shows that as the number of rectangles $n$ increases, the approximate area computed using the Left Riemann Sum gets closer and closer to the exact area under the curve, which is 
$\frac{1}{3} \approx 0.3333$

- When $n=5$ or $n=10$, the approximation is visibly below the exact value, showing a relatively large error.
- As $n$ increases to $100$ and $1000$, the approximate value approaches the exact integral, with the error becoming very small.
- the red dashed line (representing the exact area) becomes almost indistinguishable from the approximation when $n \geq 100$

This demonstrates that the Left Riemann Sum becomes more accurate as the number of rectangles increases, and that Riemann sums converge to the true value of the definite integral as $n \to \infty$

The code below is the interactive tool helps us visualize how we can approximate the area under the curve of the function $f(x) = x^2$ using different Riemann sum methods.

You can:

- Select the approximation method: Left Endpoints, Right Endpoints, or Center (Midpoint) Points

- Adjust the number of rectangles (n) used in the approximation

As you increase the number of rectangles, you'll notice that the estimated area gets closer to the actual area under the curve. This demonstrates the idea of convergence — that Riemann sums get more accurate as the partition becomes finer.

This is a powerful way to see how numerical integration works and how the shape of rectangles changes depending on the method chosen. It's also a great stepping stone toward understanding definite integrals as the limit of these approximations.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from ipywidgets import interact, Dropdown, BoundedIntText
%matplotlib inline

# Use the fivethirtyeight style
matplotlib.style.use('fivethirtyeight')

# Function f(x) = x^2
def f(x):
    return x**2

def basic_riemann(approx_type, num_rects):
    fig, ax = plt.subplots(figsize=(10, 5))  # figure size
    xs = np.linspace(0, 2.5, 300)
    ax.plot(xs, f(xs), 'darkgreen', linewidth=2, label=r"$f(x) = x^2$")

    rectangle_width = 2.5 / num_rects
    rectangle_xs = np.arange(0, 2.5, rectangle_width)
    area_approx = 0

    for x in rectangle_xs:
        if approx_type == 'Left Endpoints':
            y = f(x)
        elif approx_type == 'Right Endpoints':
            y = f(x + rectangle_width)
        elif approx_type == 'Center Points':
            y = f(x + rectangle_width / 2)
        else:
            continue  # Skip unsupported options

        ax.add_patch(Rectangle((x, 0), rectangle_width, y, alpha=0.5, facecolor='skyblue'))
        area_approx += rectangle_width * y

    ax.set_title(f"{approx_type} Approximation with {num_rects} Rectangles", fontsize=14)
    ax.set_xlabel("x", fontsize=12)
    ax.set_ylabel("f(x)", fontsize=12)
    ax.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    print("Approximated Area:", area_approx)

    # True area under x^2 from 0 to 2.5 is (2.5)^3 / 3
    actual_area = (2.5**3) / 3
    print("Computed Actual Area:", actual_area)
    print("Error (Approx - Actual):", area_approx - actual_area)

approx_type_dropdown = Dropdown(
    options=['Left Endpoints', 'Right Endpoints', 'Center Points'],
    description='Method:'
)
num_rects_input = BoundedIntText(
    value=10,
    min=1,
    max=100,
    step=1,
    description='Rectangles:'
)

interact(basic_riemann, approx_type=approx_type_dropdown, num_rects=num_rects_input);
```
