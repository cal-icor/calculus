<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">1. The Mean Value Theorem for Integrals</h1>

<h3>Why do we need the mean value theorem for integrals?</h3>

When we study the **average value** of a functin over an interval, a natural question arises: Is there at least one point where the function actually reaches its average value?

**The Mean Value Theorem for Integrals** answers this Just as the Mean Value Theorem for Derivatives guarantees a point where the instantaneous rate of change matches the average rate of change, this theorem ensures that a continuous function reaches its average value at some point in the interval.

We know that the **area under the curve** of a continuous function $f(x)$ from $x=a$ to $x=b$ given by the **define integral**: $\int_{a}^{b} f(x)dx$ (picture on the left side). Now, imagine drawing a **rectangle** that spans the same width from $a$ to $b$ (picture on the right side). The height of this rectangle is chosen so that its area is exactly the same as the area under the curve. The **Mean Value Theorem for Integrals** tell us that: there is **at least one point** $c \in [a,b]$ such that the **height of the rectangle** - which is $f(c)$-multiplied by the width $(b-a)$, gives the same area as under the curve.

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition: The Mean Value Theorem for Integrals</h2>
  <div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>  
    
   Let $f(x)$ be continuous on the closed interval $[a,b]$. Then there exsits at least one number $c \in [a,b]$ such that:
    
   $$\int_{a}^{b} f(x)dx = f(c)(b-a)$$
    
In other words, the area under the curve $f(x)$ over $[a,b]$ is the same as the area of a rectangle with $(b-a)$ and height $f(c)$, for some $c \in [a,b]$
    
Equivanlently, the average value of the function is:
    
$$f_{avg} = \frac{1}{b-a} \int_{a}^{b} f(x)dx$$
    
and the theorem states that there exists $c \in [a,b]$ such that:
    
$$f(c) = f_{avg}$$
</div>


<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2> Example #1</h2>

**Find the average value of the function $f(x) = x^2$ over the interval $[1,3]$ and find $c$ such that $f(c)$ equals the average value of the function over $[1,3]$** 

<h2> Solution </h2>

1. Compute the average value: 

$$f_{\text{avg}} = \frac{1}{3 - 1} \int_1^3 x^2 \, dx 
= \frac{1}{2} \left[ \frac{x^3}{3} \right]_1^3 
= \frac{1}{2} \left( \frac{27}{3} - \frac{1}{3} \right) 
= \frac{1}{2} \cdot \frac{26}{3} 
= \frac{13}{3}$$

2. Find $c$ such that $f(c) = \frac{13}{3}$:

$$f(c) = c^2 = \frac{13}{3} \implies \sqrt{\frac{13}{3}} \approx 2.08$$

So the function $f(x) = x^2$ equals its average value at $x = \sqrt{\frac{13}{3}} \approx 2.08$ in $[1,3]$


<h2> Python </h2>

Let's explore the Python solution for this problem by using sympy and plot the graph by using mathplotib

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol and function
x = sp.Symbol('x')
f = x**2

# Define the interval
a, b = 1, 3

# Compute the definite integral
integral = sp.integrate(f, (x, a, b))
f_avg = integral / (b - a)
c = sp.solve(f - f_avg, x)

print(f"Average value of f(x) on [{a}, {b}]: {f_avg}")
print(f"c such that f(c) = f_avg: {[float(ci) for ci in c if a <= ci <= b]}")

# Plotting
f_lam = sp.lambdify(x, f)
x_vals = np.linspace(a, b, 400)
y_vals = f_lam(x_vals)

plt.plot(x_vals, y_vals, label='$f(x) = x^2$', color='blue')
plt.hlines(float(f_avg), a, b, colors='red', linestyles='--', label=f'Average Value = {float(f_avg):.2f}')
plt.fill_between(x_vals, y_vals, alpha=0.2)
plt.axvline(float(c[0]), color='green', linestyle=':', label=f'$x = {float(c[0]):.2f}$')
plt.legend()
plt.title('Mean Value Theorem for Integrals')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">2. Fundamental Theorem of Calculus Part 1: Integrals and Antiderivatives</h1>

When we first learn about **integrals**, they are defined as limits of **Riemann sums** - a way to compute the area under a curve. Separately, we learn about antiderivative: functions whose derivative give back the original function. But what's the connection between these two powerful ideas? 

**The Fundamental Theorem of Calculus, Part 1** establishes the relationship between differentiation and integration. 

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Fundamental Theorem of Calculus, Part 1</h2>
    
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>

Let $f$ be continuous on an interval $[a,b]$, and define a function: 
    
$$F(x) = \int_{a}^{x} f(t) dt$$
    
Then $F$ is differentiable on $(a,b)$, and:
    
$$F'(x) = f(x)$$
 
</div>


<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2>Example #2.1</h2>

Use the Fundamental Theorem of Calculus, Part 1 to find the derivative of 

$$g(x) = \int_{1}^{3} \frac{1}{t^3+1}dt$$

<h2> Solution </h2>

By the Fundamental Theorem of Calculus Part 1:

$$F'(x) = \frac{d}{dx}\left(\int_{0}^{x}\frac{1}{t^3+1}dt\right) = \frac{1}{x^3+1}$$

<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2>Example #2.2</h2>

Use the Fundamental Theorem of Calculus, and the Chain rule to calculate derivatives:

$$F(x) = \int_{1}^{\sqrt{x}} sintdt$$

<h2> Solution </h2>

Letting $u(x) = \sqrt{x}$, we have $F(x) = \int_{1}^{u(x)} sint dt$. Using the Fundamental Theorem of Calculus and the chain rule:

$$F'(x) = sin(u(x)) \frac{du}{dx}$$

$$= sin(u(x)).\left(\frac{1}{2}x^{\frac{-1}{2}}\right)$$

$$= \frac{sin\sqrt{x}}{2\sqrt{x}}$$

<div style="width: 100%; height: 2px; background-color: #2c5364; margin: 20px 0;"></div>

<h2>Example #2.3</h2>

Use the Fundamental Theorem of Calculus with Two Variable Limits of Integration:

Let $F(x) = \int_{x}^{2x}t^3dt$. Find $F'(x)$.

<h2> Solution </h2>

Slipt $F(x)$ into two integrals: $F(x) = \int_{x}^{2x}t^3dt$

$$= \int_{x}^{0}t^3dt + \int_{0}^{2x} t^3dt$$

$$= - \int_{0}^{x}t^3dt + \int_{0}^{2x} t^3dt$$.

For the first term:

$$\frac{d}{dx}\left[\int_{0}^{2x} t^3dt\right]= -x^3$$.

For the second term, let $u(x) = 2x$:

$$\frac{d}{dx}\left[\int_{0}^{2x} t^3dt\right] = \frac{d}{dx}\left[\int_{0}^{u(x)} t^3dt\right] $$

$$= (u(x))^3\frac{du}{dx}$$

$$=(2x)^3.2$$

$$=16x^3$$.

Thereofore:

$$F'(x) = \frac{d}{dx}\left[\int_{0}^{2x} t^3dt\right] + \frac{d}{dx}\left[\int_{0}^{2x} t^3dt\right]$$

$$= -x^3 + 16x^3$$

$$= 15x^3$$.

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">3. Fundamental Theorem of Calculus Part 2: The Evaluation Theorem</h1>

Let’s revisit how we define definite integrals using limits of Riemann sums. That’s a great theoretical foundation, but it’s computationally inefficient. For example, to compute:

$$\int_{1}^{4} (2x+1)dx$$

Do we really want to approximate this using many rectangles and a limit? Not really! Especially when we already know how to take antiderivatives!

That's where the Fundamental Theorem of Calculus, Part 2 comes in. Instead of computing an integral using sums, we can evaluate any definite integral by simply finding an antiderivative of the function and evaluating it at the endpoints.

It turns integration into a much easier process — and it explains why antiderivatives are so important.

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Fundamental Theorem of Calculus, Part 2</h2>
    
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>

Let $f$ be continuous on an interval $[a,b]$, and let $F$ be any antiderivative of $f$, meaning $F'(x) = f(x)$. Then: 
    
$$\int_{a}^{b} f(x) dx= F(b) - F(a)$$
    
 
</div>


<h2> Example #3.1 </h2>

Use the Fundamental Theorem of Calculus Part 2 to evaluate the integral:

$$\int_{1}^{4} \left(\frac{1}{x^2}\right)dx$$

<h2> Solution </h2>

We know the antiderivative of $\frac{1}{x^2}$ is $-\frac{1}{x}$

Apply the Fundamental Theorem of Calculus Part 2:

$$\int_{1}^{4} \left(\frac{1}{x^2}\right)dx = \left[-\frac{1}{x}\right]_1^4 = -\frac{1}{4} - \left(-\frac{1}{1}\right)= \frac{3}{4}$$

<h2> Python </h2>

Let's explore the solution in Python for this problem. Luckily, we have the powerful **integrate()** function from the SymPy library. **integrate(f, (x, a, b))** helps us compute the definite integral of the function $f$ with respect to $x$ from $a$ to $b$.

```python
import sympy as sp

# Define the variable
x = sp.Symbol('x')

# Define the function
f = 1 / x**2

# Compute the definite integral from 1 to 4
result = sp.integrate(f, (x, 1, 4))

print("∫₁⁴ 1/x² dx =", result)
```

<h2 style="color: #2c5364;"> The tool to view the area between the curve $f(x) = x^2$ and the x-axis </h2>

This interactive visualization demonstrates how we calculate the area under a curve using definite integrals. The function we're working with is $f(x) = x^2$, and the shaded region on the graph represents the area between the curve and the x-axis over a selected interval $[a,b]$

Use the sliders labeled “Lower Bound (a)” and “Upper Bound (b)” to select your interval.

As you adjust the bounds, the graph will update in real time:

- The blub curve shows the graph of $f(x) = x^2$.

- The shaded area shows the region under the curve from $x=a$ to $x=b$.

Below the graph, you'll see the calculated area using numerical integration.

This is a helpful way to visualize how integration works and how the area depends on the interval you choose.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.integrate import quad
%matplotlib inline

# Define the function
def f(x):
    return x**2

# Plotting function with adjustable bounds
def plot_integral_area(a, b):
    if a > b:
        a, b = b, a  # Ensure a < b for integration

    x = np.linspace(-1, 3, 300)
    y = f(x)

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(x, y, label=r"$f(x) = x^2$", color='darkblue', linewidth=2)

    # Shade the area under the curve between a and b
    x_fill = np.linspace(a, b, 300)
    y_fill = f(x_fill)
    ax.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label='Area under curve')

    # Compute the integral value
    area, _ = quad(f, a, b)

    # Labels and annotations
    ax.set_title("Area Under $f(x) = x^2$ from a to b", fontsize=14)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.set_xlim([-1, 3])
    ax.set_ylim([0, 10])
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    print(f"Approximate area under the curve from x = {a} to x = {b}: {area:.4f}")

# Sliders for bounds of integration
a_slider = FloatSlider(value=0.0, min=-1, max=3, step=0.1, description='Lower Bound (a)')
b_slider = FloatSlider(value=1.0, min=-1, max=3, step=0.1, description='Upper Bound (b)')

interact(plot_integral_area, a=a_slider, b=b_slider);
```

3. **Apply the Fundamental Theorem of Calculus (Part 2)**: Evaluate the antiderivative at the endpoints and subtract:
    
$$\int_{a}^{b} f(x) dx= F(b) - F(a)$$
    
4. **Simplify the result**: Perform any necessary arithmetic or simplification to get the final answer.

<h2> Example #3.2 </h2>

Use the Fundamental Theorem of Calculus Part 2 to evaluate the integral:

$$\int_{-1}^{2} sin(x)dx$$

<h2> Solution </h2>

Apply the Fundamental Theorem of Calculus Part 2:

$$\int_{-1}^{2} sin(x)dx = \left[-cos(x)\right]_{-1}^2 = -cos(2)+cos(1) \approx 0.9564$$

<h2> Python </h2>

```python
import sympy as sp

# Step 1: Define the variable
x = sp.Symbol('x')

# Step 2: Define the function f(x)
f = sp.sin(x)

# Step 3: Define the bounds
a = -1
b = 2

# Step 4: Find the antiderivative (optional if just using integrate)
F = sp.integrate(f, x)  # Indefinite integral

# Step 5: Evaluate the definite integral
result = F.subs(x, b) - F.subs(x, a)


# Step 6: Output the result
print("Antiderivative F(x) =", F)
print(f"Definite integral from {a} to {b} is:", result)
```

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">4. Real-World Applications</h1>

Let’s return to the real-world motivation we introduced at the beginning: using integration to find the distance traveled when we know the velocity over time.

In everyday life, velocity isn’t just a mathematical function — it's a real, measurable quantity. If we know how fast something is moving over time, we can calculate how far it goes by integrating that velocity function.

Let's say we have a velocity curve and what if we were driving at a constant speed $v(t) = 10$ $(m/s)$ from $t=3$ to $t=8$ seconds.

We can see that the area under that curve would be the distance. We know that $s'(t) = v(t)$, we could find that distance by using the Fundamental Theorem of Calculus Part 2:

$$\int_{3}^{8}10 = \left[10t\right]_3^8 = 80 - 30 = 50 (meters)$$ 

But unfortunately (or realistically), we don’t usually travel at a constant speed. Traffic, stop signs, and changes in terrain cause our speed to change.

For example, consider a more realistic velocity function: 

$$v(t) = t(8-t)$$



This function starts at $0$, increases to a maximum, and then decreases — forming a parabolic curve. The area under this curve from $t=0$ to $t=8$ gives the total distance traveled:

$$D = \int_0^8 t(8-t)dt$$

This is where the **Fundamental Theorem of Calculus** becomes truly powerful: it allows us to find the total distance without approximating rectangles or counting squares — instead, we find an antiderivative and evaluate it efficiently.

$$D = \int_0^8 t(8-t)dt = \int_0^8 (8t-t^2)dt = [4t^2-\frac{1}{3}t^3]_3^8 = [4(8)^2-\frac{1}{3}(8)^3] - [4(3)^2-\frac{1}{3}(3)^3] = \frac{175}{3} (meters)$$

$$Volume = \int_a^b r(t)dt$$
    
This is common in environmental science, plumbing design, and fluid mechanics.
