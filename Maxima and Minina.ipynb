<h1> Introduction </h1>

In calculus, we often want to find the highest or lowest points of a function. These are called maximum and minimum values, or extrema. These are important in creating accurate graphs. In real life, maxima and minima help answer questions such as: What price will maximize profit? What path minimizes travel time? What dimensions give the smallest surface area for a given volume? Understanding extrema allows us to optimize processes in physics, economics, engineering, biology, and more. In this section, we will explore how to use derivatives to find largest and smallest values for a function.

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">1. Absolute Extrema</h1>

Consider the function $f(x) = 4-x^2$ over the interval $(-\infty, \infty)$. This is a classic example of a downward-opening parabola, which forms a smooth, symmetrical curve shaped like an upside-down bowl (see picture below).

As you move from left to right:

- The function increases until it reaches the very top at $x=0$.

- The it decreases as you keep moving right.

![Screenshot%202025-06-25%20at%201.17.20%E2%80%AFPM.png](images/max_min1.png)

Just by looking at the graph, we can see that the highest point on the curve is at $x=0$, where $f(0)=4$. This point is called the absolute maximum because it's the largest value the function ever reaches.

As you move far left or far right (toward $x = -\infty$ or $x=\infty$), the values of $f(x)$ decrease forever — the function never bottoms out at one single lowest point. That means it doesn’t have an absolute minimum on the entire real number line.

This function is a great way to visualize the idea of an absolute maximum, even if we don’t use derivatives yet.

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition: Absolute Extrema</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>
    
Let $f$ be defined on a domain $D$.
    
- $f(c)$ is the absolute maximum if $f(c) \geq f(x)$ for all $x \in D$.
    
- $f(c)$ is the absolute minimum if $f(c) \leq f(x)$ for all $x \in D$.
    
$f$ has a absolute extremum at $c$ if $f$ has a local maximum at $c$ or $f$ has a local minimum at $c$.
</div>


![Screenshot%202025-06-20%20at%209.39.05%E2%80%AFPM.png](images/max_min2.png)

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition: Extreme Value Theorem</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>

If a function is **continuous** on a **closed interval** $[a,b]$, then it **must** attain both an absolute maximum and minimum value on that interval.
</div>

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">2. Local Extrema and Critical Points</h1>

In Section 1, we explored **absolute extrema**—the highest and lowest values a function reaches over an entire domain or a specific interval. However, not all important turning points are absolute. Many real-world scenarios require us to examine a function's behavior locally, or in a small neighborhood around a point. These local high and low points are called **local extrema** (local maximums and minimums). A local maximum occurs when the function value at a point is higher than the values of nearby points. A local minimum is when it's lower than all nearby values.

Consider the function $f$ shown in the picture below. As we mentioned in section 1, we know that it has **absolute maximum** at $x=0$. However, there are some point of interest, such as the peaks of mountains and the bottoms of valleys:

- At $x=3$, the value $f(3)$ is smaller than $f(x)$ for all $x$ near $3$, we can say $f$ has the **local minimum** at $x=3$.

- Similarly, at $x=4$, $f$ also has the **local maximum** because $f(4)$ is greater than $f(x)$ for all $x$ near $4$.

- $f$ has an **absolute maximum** at $x=0$ and this also is a **local maximum** for all $x$ around $0$.

![Screenshot%202025-06-23%20at%2010.06.38%E2%80%AFAM.png](images/max_min3.png)

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition: Local Extrema</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>
    
Let $f$ be defined on a domain $D$ and let $c \in D$
    
- Local maximum at $c$ if $f(c) \geq f(x)$ for all $x$ near $c$
    
- Local minimum at $c$ if $f(c) \leq f(x)$ for all $x$ near $c$.
    
$f$ has a local extremum at $c$ if $f$ has a local maximum at $c$ or $f$ has a local minimum at $c$.
</div>


As we mention on the graph of $f$ above, $f$ has an absolute maximum at $0$ and this is also a local maximum on the interval that includes $0$.

If a function $f$ has an absolute extremum at a point $c$, and $f$ is defined on an interval that includes $c$, $f(c)$ is also considered a local extremum. However, if the absolute extremum happens at an endpoint, it is not labeled as a local extremum—it is simply referred to as an endpoint extremum.

Sometimes we can visually identify local maxima or minima by looking at a graph. But this isn’t always reliable, especially when important features occur at a small scale or when no graph is available. In those cases, we rely on formulas and calculus tools to locate these points.

For example, a graph might show local extrema at $x=0$, $x=3$, and $x=4$. At $x=0$ and $x=3$, the derivative $f'(x) = 0$, indicating possible extrema. At $x=4$, $f'(x)$ does not exist because the function has a sharp corner there.

This leads to a key idea: if a function has a local extremum at a point $x=c$, then either:

- $f'(c) = 0$, or

- $f'(x)$ is **undefined**.

such a point $c$ is called a **critical point**, and identifying critical points is crucial for finding extrema values of a function.

![Screenshot%202025-06-23%20at%2010.36.20%E2%80%AFAM.png](images/max_min4.png)

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Definition: Critical Points</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>
    
A number $c$ in the domain of $f$ is a critical number if:
    
- $f'(c) = 0$, or

- $f'(x)$ does not exist.
</div>


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Fermat's Theorem</h2>
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>
    
If $f$ has a local extremum at $c$ and $f$ is differentiable at $c$, then $f'(c) = 0$.
</div>

<h2> Identify critical points graphically </h2>
    
As a result, to determine critical points of a function, we simply check where the derivative is either equal to zero or does not exist (see the pictures below)

- Case $f'(c) = 0$, there is a **horizontal tangent line** at $x=c$.

- Case $f'(c)$ is undefined:

    - There is a **sharp corner** at $c$: The graph changes direction abruptly and there is no well-defined tangent line..
    
    - There is a **vertical tangent line****: the derivative approaches infinity as we get close to $c$
    
    - $f$ is **not continuous** at $c$**: the derivative cannot exist there.

<h2> Example #2.1 </h2>

**Consider function $f(x) = x^3 - 3x$.**

a) Find all critical points

b) Using Python to find all critical points and graph the function $f(x)$

c) Use the graph is created in part b to determine whether the function has a local extremum at each of the critical points.

<h2> Solution </h2>

For part (a):

Find the derivative of $f(x)$ and set it to zero: $f'(x) = 3x^2 - 3=0$.

Critical points at $x=-1$ and $x=1$

<h2> Python </h2>

For part (b), use diff(f,x) to find derivative of $f$ with respect to $x$. 

Then using solve(f,x) to solve $f'(x) = 0$

```python
import sympy as sp
import numpy as np

# Symbolic variable
x = sp.Symbol('x')

# Function f(x) = x^3 - 3x
f = x**3 - 3*x

# Find the derivative f'(x)
f_prime = sp.diff(f, x)

# Find critical points: solve f'(x) = 0
crit_points = sp.solve(f_prime, x)

crit_points_vals = [(cp, f.subs(x, cp)) for cp in crit_points]

# Print results
print("Critical points:")
for x_vals in crit_points:
    print(x_vals)
    
    
print("-------------")

print("Critical points values:")
for cp, val in crit_points_vals:
    print(f"f({cp}) = {val}")
```

The code below shows that how we use matplotlib and numpy for graphing

```python
import matplotlib.pyplot as plt

x_vals = np.linspace(-4, 4, 400)

# Plot f(x) = x^3 - 3x
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
f1_np = lambda x: x**3 - 3*x
plt.plot(x_vals, f1_np(x_vals), label=r'$f(x) = x^3 - 3x$')
plt.scatter(*zip(*crit_points_vals), color='red', label='Critical Points')
plt.title('Critical Points of $f(x) = x^3 - 3x$')
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

For part (c), as the graph shows above:

- $f$ has local maximum at $x=-1$

- $f$ has local minimum at $x=1$.

<h2> Example #2.2 </h2>

**Consider function $f(x) = (x^2-1)^3$.**

a) Find all critical points

b) Using Python to find all critical points and graph the function $f(x)$

c) Use the graph is created in part b to determine whether the function has a local extremum at each of the critical points.

<h2> Solution </h2>

For part (a): Using the chain rule to find the derivative of the function:

$$f'(x) = 3(x^2-1)^2(2x) = 6x(x^2-1)^2$$

Set it to zero and we get the critical points when $x=0, \pm 1$

<h2> Python </h2>

For part (b): Similarly the previous example, press Shift and Enter at the same time to run the code below to see the result:

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Symbolic variable
x = sp.Symbol('x')

# Function: f(x) = (x^2 - 1)^3
f = (x**2 - 1)**3

# Derivative
f_prime = sp.diff(f, x)

# Solve f'(x) = 0 to find critical points
crit_points = sp.solve(f_prime, x)

# Evaluate function at each critical point
crit_points_vals = [(cp, f.subs(x, cp)) for cp in crit_points]

# Print critical numbers
print("Critical points:")
for cp in crit_points:
    print(cp)

print("-------------")

# Print critical points and their values
print("Critical points values:")
for cp, val in crit_points_vals:
    print(f"f({cp}) = {val}")

# Prepare to plot the function
f_np = sp.lambdify(x, f, 'numpy')  # Convert to NumPy-compatible function
x_vals = np.linspace(-2, 2, 400)
y_vals = f_np(x_vals)

# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = (x^2 - 1)^3$')
plt.scatter(
    [float(cp) for cp, _ in crit_points_vals],
    [float(val) for _, val in crit_points_vals],
    color='red',
    label='Critical Points'
)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.title('Critical Points of $f(x) = (x^2 - 1)^3$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

As we see the graph above:

$f$ has a local minimum at $x=0$, but does not have a local extremum at $x=1$ or $x=-1$ because it doesn't switch from increasing to decreasing or vice versa at those points.

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">3. Locating Absolute Extrema</h1>

The Extreme Value Theorem guarantees that if a function is continuous on a closed and bounded interval, it will attain both an absolute maximum and an absolute minimum on that interval. These extreme values may occur at the endpoints or somewhere inside the interval. If they occur inside, they are also considered local extrema. According to Fermat’s Theorem, if an absolute extremum occurs at an interior point, that point must be a critical point (where the derivative is zero or undefined). This gives us a structured method to find absolute extrema by checking both the endpoints and the critical points.

<a id="strategy"></a>
<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  <h2 style="margin:0;">Strategy: Locating Absolute Extrema over a Closed Interval</h2>
    
<div style="width: 100%; height: 2px; background-color: #ccc; margin: 20px 0;"></div>

Consider a continuous functin $f$ defined over the closed interval $[a,b]$.
    
1. Find the derivative $f'(x)$

2. Solve $f'(x) = 0$ and find critical points inside $[a,b]$
    
3. Evaluate the function $f(x)$ at:
    
    - Each critical point within $[a.b]$
    
    - The endpoints $a$ and $b$.

4. Compare all these values:

    - The largest value is the absolute maximum

    - The smallest value is the absolute minimum

<h2> Example #3.1 </h2>

Find the absolute maximum and absolute minimum of the function:

$$f(x) = x^2$$

on the interval $[-1,2]$

<h2> Solution </h2>

Follow the strategy above:

1. Find the derivative: 
    
    $$f'(x) = 2x$$
    
2. Set it to zero and find critical points:

$$f'(x) = 2x = 0 \implies x = 0$$

3. Evaluate the function $f(x)$ at:

- Critical point: 

$$x=0 \implies f(0) = (0)^2 = 0$$

- Endpoints:

$$x=-1 \implies f(-1) = (-1)^2 = 1$$

$$x = 2 \implies f(2) = (2)^2 = 4$$

4. Compare all these values:

\begin{array}{|c|c|l|}
\hline
x & f(x) & \text{Conclusion} \\
\hline
-1 & 1 & \\
0 & 0 & \text{Absolute Minimum} \\
2 & 4 & \text{Absolute Maximum} \\
\hline
\end{array}

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Step 0: Define symbol and function
x = sp.Symbol('x')
f = x**2

# Step 1: Find derivative and critical points
f_prime = sp.diff(f, x)
crit_points = sp.solve(f_prime, x)

# Keep only critical points within the interval [-1, 2]
a, b = -1, 2
crit_points_in_interval = [cp for cp in crit_points if a <= cp <= b]

# Step 3: Evaluate f(x) at critical points and endpoints
eval_points = crit_points_in_interval + [a, b]
eval_vals = [(float(pt), float(f.subs(x, pt))) for pt in eval_points]

# Step 4: Find absolute min and max
min_point = min(eval_vals, key=lambda t: t[1])
max_point = max(eval_vals, key=lambda t: t[1])

# Display results
print("Evaluations:")
for pt, val in eval_vals:
    print(f"f({pt}) = {val}")

print(f"\nAbsolute Minimum: f({min_point[0]}) = {min_point[1]}")
print(f"Absolute Maximum: f({max_point[0]}) = {max_point[1]}")

# Plotting the function and the extrema
f_np = sp.lambdify(x, f, 'numpy')
x_vals = np.linspace(a - 0.5, b + 0.5, 400)
y_vals = f_np(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, color='blue')
plt.scatter(*zip(*eval_vals), color='red', label='Evaluated Points', zorder=5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

- Critical point: 

$$x=\frac{\pi}{2} \implies f\left(\frac{\pi}{2}\right) = sin\left(\frac{\pi}{2}\right) = 1$$

$$x= 0 \implies f(0) = sin(0) = 0$$

$$x = 2\pi \implies f(2\pi) = sin(2\pi) = 0$$

<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">4. Real-World Applications</h1>

Let’s return to the real-world motivation we introduced at the beginning: using derivatives to find the highest or lowest values of a function.

In everyday life, these maximum and minimum values aren't just abstract math — they help us make real decisions. For example, businesses want to maximize profit or minimize cost, engineers want to maximize efficiency, and travelers want to minimize time.

By using calculus, we can take a function that models a real situation and use its derivative to find the point where something is largest or smallest. This is called optimization, and it’s one of the most powerful tools calculus gives us.

<h2> Example: Maximizing Profit </h2>
    
A company models its profit (in thousands of dollars) as a function of the number of items produced: 

$$P(x) = -2x^2 + 40x -120$$

where $x$ is the number of items (in hundreds). Find the number of items that maximizes profit

<h2> Solution </h2>

Find the derivative:

$$P'(x) = -4x+40$$

Set it to zero:

$$-4x+40 = 0 \implies x=10$$

So, the company maximizes its profit by producing **1,000 items**.
