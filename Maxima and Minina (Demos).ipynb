## 1. Absolute extrema

**Key idea:** On a closed interval $[a,b]$, a continuous function attains an **absolute maximum** and **absolute minimum** (Extreme Value Theorem). These are the highest and lowest values of $f$ on $[a,b]$.

**Use the demo below.** Change the interval $[a,b]$ and the function. Identify where the absolute max and min occur.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown
%matplotlib inline

def abs_extrema_demo(func_choice='4 - x^2', a=-2, b=2):
    if func_choice == '4 - x^2':
        def f(x): return 4 - x**2
        x_plot = np.linspace(-3, 3, 300)
    elif func_choice == 'x^3 - 3x':
        def f(x): return x**3 - 3*x
        x_plot = np.linspace(-2.5, 2.5, 300)
    else:
        def f(x): return 4 - x**2
        x_plot = np.linspace(-3, 3, 300)
    
    a, b = min(a, b), max(a, b)
    x = np.linspace(a, b, 200)
    y = f(x)
    y_max, y_min = np.max(y), np.min(y)
    x_max = x[np.argmax(y)]
    x_min = x[np.argmin(y)]
    
    plt.figure(figsize=(9, 5))
    plt.plot(x_plot, f(x_plot), 'b-', lw=2, label=f'$f(x)$, {func_choice}')
    plt.fill_between(x, f(x), alpha=0.2)
    plt.axvline(a, color='gray', linestyle='--', alpha=0.7, label=f'$a={a}$')
    plt.axvline(b, color='gray', linestyle='--', alpha=0.7, label=f'$b={b}$')
    plt.plot(x_max, y_max, 'ro', markersize=12, label=f'Abs max: $f({x_max:.2f})={y_max:.3f}$')
    plt.plot(x_min, y_min, 'go', markersize=12, label=f'Abs min: $f({x_min:.2f})={y_min:.3f}$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend(loc='best', fontsize=9)
    plt.title(f'Absolute extrema on $[{a}, {b}]$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

interact(abs_extrema_demo,
         func_choice=Dropdown(options=['4 - x^2', 'x^3 - 3x'], value='4 - x^2', description='f(x)'),
         a=FloatSlider(value=-2, min=-3, max=2, step=0.25, description='a'),
         b=FloatSlider(value=2, min=-2, max=3, step=0.25, description='b'));
```

## 2. Local extrema and critical points

**Key idea:** A **critical point** is where $f'(c)=0$ or $f'$ is undefined. Local extrema occur at critical points (or endpoints).

**Demo 1 (below):** Choose a function and see the graphs of $f$ and $f'$. Critical points are where $f'$ crosses zero (red vertical lines).

**Demo 2 (further below):** Use the slider to **place the critical point** yourself on the $x$-axis and get feedback on how close you are.

```python
def critical_points_demo(func_choice='x^3 - 3x'):
    if func_choice == 'x^3 - 3x':
        def f(x): return x**3 - 3*x
        def fp(x): return 3*x**2 - 3
        critical = [-1.0, 1.0]
        x_range = (-2.5, 2.5)
    elif func_choice == '4 - x^2':
        def f(x): return 4 - x**2
        def fp(x): return -2*x
        critical = [0.0]
        x_range = (-3, 3)
    else:
        def f(x): return x**3 - 3*x
        def fp(x): return 3*x**2 - 3
        critical = [-1.0, 1.0]
        x_range = (-2.5, 2.5)
    
    x = np.linspace(x_range[0], x_range[1], 300)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 7), sharex=True)
    ax1.plot(x, f(x), 'b-', lw=2, label=r'$f(x)$')
    for c in critical:
        ax1.plot(c, f(c), 'ro', markersize=10)
        ax1.axvline(c, color='red', linestyle=':', alpha=0.6)
    ax1.set_ylabel('$y$')
    ax1.legend()
    ax1.set_title(f'$f(x)$ and critical points at $x = {", ".join(str(round(c, 1)) for c in critical)}$')
    ax1.grid(True, alpha=0.3)
    ax2.plot(x, fp(x), 'g-', lw=2, label=r"$f'(x)$")
    ax2.axhline(0, color='k', linestyle='-', lw=0.5)
    for c in critical:
        ax2.axvline(c, color='red', linestyle=':', alpha=0.6)
    ax2.set_xlabel('$x$')
    ax2.set_ylabel(r"$f'(x)$")
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

interact(critical_points_demo,
         func_choice=Dropdown(options=['x^3 - 3x', '4 - x^2'], value='x^3 - 3x', description='f(x)'));
```

**Place the critical point:** Use the slider to mark where you think a critical point is on the x-axis. You'll get feedback on how close you are.

```python
def place_critical_point(guess_x=0.0):
    # For f(x) = x^3 - 3x, critical points at x = -1 and x = 1
    def f(x): return x**3 - 3*x
    def fp(x): return 3*x**2 - 3
    actual = [-1.0, 1.0]
    
    x = np.linspace(-2.5, 2.5, 300)
    plt.figure(figsize=(8, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x)=x^3-3x$')
    plt.axhline(0, color='k', lw=0.5)
    for c in actual:
        plt.plot(c, f(c), 'go', markersize=10, label=f'Actual critical: $x={c}$')
    plt.axvline(guess_x, color='red', linestyle='--', linewidth=2, label=f'Your placement: $x={guess_x}$')
    plt.plot(guess_x, f(guess_x), 'rs', markersize=10)
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend(loc='best', fontsize=9)
    plt.title('Place the critical point (Option B: slider)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Feedback
    dist = min(abs(guess_x - c) for c in actual)
    if dist < 0.05:
        print('Correct! You placed a critical point very close to x = -1 or x = 1.')
    else:
        print(f'You placed the critical point at x = {guess_x}. The actual critical points are at x = -1 and x = 1. Try moving the slider.')

interact(place_critical_point,
         guess_x=FloatSlider(value=0.0, min=-2, max=2, step=0.1, description='Your x'));
```

## 3. Locating absolute extrema

**Key idea:** On $[a,b]$, the absolute max and min occur at **critical points** (in $(a,b)$) or at **endpoints** $a$ and $b$. So: find critical points, evaluate $f$ at each and at $a$, $b$; the largest value is the max, the smallest is the min.

**Use the demo below.** Change $[a,b]$ and see the candidate points and the resulting max/min.

```python
def locate_extrema_demo(a=-2, b=2):
    def f(x): return x**3 - 3*x
    def fp(x): return 3*x**2 - 3
    a, b = min(a, b), max(a, b)
    
    # Critical points in (a,b) where f'(x)=0
    crit = [x for x in [-1.0, 1.0] if a < x < b]
    candidates = [a] + crit + [b]
    candidates = sorted(set(candidates))
    values = [f(x) for x in candidates]
    imax = np.argmax(values)
    imin = np.argmin(values)
    x_max, y_max = candidates[imax], values[imax]
    x_min, y_min = candidates[imin], values[imin]
    
    x = np.linspace(a - 0.5, b + 0.5, 300)
    plt.figure(figsize=(9, 5))
    plt.plot(x, f(x), 'b-', lw=2, label=r'$f(x)=x^3-3x$')
    plt.axvline(a, color='gray', linestyle='--', alpha=0.7)
    plt.axvline(b, color='gray', linestyle='--', alpha=0.7)
    for i, (xc, yc) in enumerate(zip(candidates, values)):
        label = 'Endpoint' if xc in (a, b) else 'Critical'
        plt.plot(xc, yc, 'ro' if xc == x_max or xc == x_min else 'ko', markersize=10)
    plt.plot(x_max, y_max, 'r*', markersize=18, label=f'Abs max: $f({x_max:.2f})={y_max:.3f}$')
    plt.plot(x_min, y_min, 'g*', markersize=18, label=f'Abs min: $f({x_min:.2f})={y_min:.3f}$')
    plt.xlabel('$x$')
    plt.ylabel('$y$')
    plt.legend(loc='best', fontsize=9)
    plt.title(f'Candidates: $x = {candidates}$')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    print(f'Candidates: x = {candidates}')
    print(f'f(candidates) = {[round(f(x), 4) for x in candidates]}')
    print(f'Absolute max: f({x_max:.4f}) = {y_max:.4f}; Absolute min: f({x_min:.4f}) = {y_min:.4f}')

interact(locate_extrema_demo,
         a=FloatSlider(value=-2, min=-2.5, max=1.5, step=0.25, description='a'),
         b=FloatSlider(value=2, min=0.5, max=2.5, step=0.25, description='b'));
```

## 4. Real-world (brief)

Derivatives help solve optimization problems: maximize profit, minimize cost, or maximize area for a fixed perimeter. The idea is the same: find critical points (and check endpoints when there is a constraint interval), then evaluate the function to find the best value.

**Optional demo:** Maximize area of a rectangle with perimeter 20: one side is $x$, the other is $10-x$, so area $A(x)=x(10-x)$. Use the slider to see how area changes.

```python
def area_demo(x=5.0):
    from IPython.display import display, Math
    # Perimeter 20: 2x + 2y = 20 => y = 10 - x. Area A(x) = x(10-x)
    x = max(0.1, min(9.9, x))
    y = 10 - x
    A = x * y
    A_max = 25.0  # at x = 5

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))

    # Left: rectangle
    rect = plt.Rectangle((0, 0), x, y, fill=True, edgecolor='blue', linewidth=2, alpha=0.5)
    ax1.add_patch(rect)
    ax1.set_xlim(0, 12)
    ax1.set_ylim(0, 12)
    ax1.set_aspect('equal')
    ax1.set_xlabel('$x$')
    ax1.set_ylabel('$y$')
    ax1.set_title(f'Rectangle: perimeter 20, sides $x$ and $10-x$\n$x = {x:.1f}$, $y = {y:.1f}$, Area = {A:.2f}')
    ax1.grid(True, alpha=0.3)

    # Right: graph of A(x) = x(10-x)
    x_plot = np.linspace(0.1, 9.9, 200)
    ax2.plot(x_plot, x_plot * (10 - x_plot), 'b-', lw=2, label=r'$A(x) = x(10-x)$')
    ax2.plot(x, A, 'ro', markersize=12, label=f'Current: $x = {x:.1f}$, $A = {A:.2f}$')
    ax2.axvline(5, color='green', linestyle='--', alpha=0.8, label=r'Max at $x = 5$')
    ax2.plot(5, 25, 'g*', markersize=14, label=r'Max area $= 25$')
    ax2.set_xlabel('$x$')
    ax2.set_ylabel('$A(x)$')
    ax2.legend(loc='upper right', fontsize=9)
    ax2.set_ylim(0, 30)
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    display(Math(r'A(x) = x(10-x) = %s \times %s = %s' % (x, "%.1f" % y, "%.2f" % A)))
    display(Math(r"A'(x) = 10 - 2x = 0 \Rightarrow x = 5 \text{ gives max area } 25"))

interact(area_demo, x=FloatSlider(value=5, min=0.5, max=9.5, step=0.5, description='x'));
```

## Summary

- **Absolute extrema:** Highest and lowest values of $f$ on a set; on a closed interval $[a,b]$, a continuous function attains both (Extreme Value Theorem).
- **Critical points:** $c$ where $f'(c)=0$ or $f'$ is undefined. Local extrema occur at critical points (or endpoints).
- **Locating absolute extrema on $[a,b]$:** Find critical points in $(a,b)$, then evaluate $f$ at those points and at $a$ and $b$; the largest value is the max, the smallest is the min.
