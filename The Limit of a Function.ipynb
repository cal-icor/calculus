<h1> Introduction </h1>

In Calculus 1, the limit of a function describes the behavior of the function as the input approaches a certain value. Rather than focusing on the function's value at that point, a limit tells us what value the function is getting closer to. Limits are essential for understanding concepts like continuity, derivatives, and the overall behavior of graphs. For example, when a function approaches a certain number from both sides of a point—even if it’s not defined exactly at that point—we can still describe its limit. This foundational idea helps build the tools used to study change and motion in calculus.

<a id="section1"></a>
<h1 style="border-left: 6px solid #2c5364; padding-left: 10px;">1. Definition of Limits</h1>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #001f4d;">
Let $f(x)$ be a function that is defined for all values near $a$, except possibly at $a$ itself. Let $L$ be a real number. If the values of $f(x)$ get closer to $L$ as $x$ (not equal to $a$) gets closer to $a$, then we say the limit of $f(x)$ as $x$ approaches $a$ is $L$. In other words, as $x$ gets closer to $a$, the function values approach and stay near $L$. This idea is written symbolically as:

$$
\displaystyle \lim_{x \to a} f(x) = L
$$


</div>

<h2>Example</h2>

Consider the function $f(x) = \frac{x^2-4}{x-2}$ behaves near $x=2$. As $x$ gets closer to $2$ from both sides, the value of $f(x)$ gets closer to $4$, even the function is not defined at $x=2$. So, mathematically, we say the limit of $f(x)$ as $x$ approaches $2$ is $4$. 

We can express this idea as:

$$
\displaystyle \lim_{x \to 2} f(x) = 4
$$

We can use the following steps to investigate any limit we seek of the form $\displaystyle\lim_{x \to a} f(x)$.

**Step 1.** Create a table of values:
    To estimate the limit of a function $\lim_{x \to a} f(x)$, we start by picking values of $x$ that get closer to $a$ from both sides of $a$ (left and right). X-values that approach $a$ from the left, are numbers that are smaller than $a$, but get closer and closer to $a$, like: $a-0.1$, $a-0.01$, etc..  If we want to choose values of $x$ that approach $a$ from the right, these are values that are larger than $a$ and get closer and closer to $a$, like: $a+0.1$, $a+0.01$, $a+0.001$, etc (use more values if needed to see the trend more clearly).
    
<div style="display: flex; justify-content: center; gap: 40px; margin-top: 30px">

<div style="margin-right: 80px;">
    

| $x$         | $\lim\limits_{x \to a^-} f(x)$         |
|-------------|------------------|
| $a - 0.1$   | $f(a - 0.1)$     |
| $a - 0.01$  | $f(a - 0.01)$    |
| $a - 0.001$ | $f(a - 0.001)$   |

</div>

<div>

| $x$         | $\lim\limits_{x \to a^+} f(x)$         |
|-------------|------------------|
| $a + 0.1$   | $f(a + 0.1)$     |
| $a + 0.01$  | $f(a + 0.01)$    |
| $a + 0.001$ | $f(a + 0.001)$   |



</div>

</div>

2. Look for a pattern:
    Observe the values of $f(x)$ as $x$ approaches $a$ from both the left and the right. If the function values in both directions seem to be getting closer to the same number, that number is likely the limit.
    
3. State the limit:
    If both sides of the table show values getting closer the the same L, we can conclude that the limit $\lim_{x \to a} f(x) = L$.
    
4. Use graphing for confirmation:
    We can also graph the function near $x=a$ using a calculator or computer. By tracing the function and observing the output values as $x$ gets closer to $a$ from both sides, you can confirm whether they are approaching the same value $L$. If needed, zoom in on the raph for better accuracy.

</div>

<h2>Example #1</h2>
Let's say we would like to find: $ \displaystyle\lim_{{x \to 2}} \frac{x^2 - 4}{x - 2}$


<h2> Solution </h2>

With $a=2$, values of $x$ on the left-hand side

| $x$ | $\displaystyle\lim_{x \to a^-} f(x)$ |
| -------- | -------- | 
| $a-0.2=1.800$ | $\lim\limits_{x \to 1.8} f(x) = 3.800$ | 
| $a-0.16=1.840$ | $\lim\limits_{x \to 1.84} f(x) = 3.840$ |
| $a-0.12=1.880$ | $\lim\limits_{x \to 1.88} f(x) = 3.880$ |
| $a-0.08=1.920$ | $\lim\limits_{x \to 1.92} f(x) = 3.920$ |
| $a-0.04=1.960$ | $\lim\limits_{x \to 1.96} f(x) = 3.960$ |

Values of $x$ on the right-hand side

| $x$ | $\displaystyle\lim_{x \to a^+} f(x)$ |
| -------- | -------- | 
| $a+0.001=4.001$ | $\lim\limits_{x \to 4.001} f(x) = 4.001$ | 
| $a+0.051=4.051$ | $\lim\limits_{x \to 4.051} f(x) = 4.051$ |
| $a+0.101=4.100$ | $\lim\limits_{x \to 4.1} f(x) = 4.100$ |
| $a+0.150=4.150$ | $\lim\limits_{x \to 4.15} f(x) = 4.150$ |
| $a+0.200=4.200$ | $\lim\limits_{x \to 4.2} f(x) = 4.200$ |

In the first table, as $x$ increases toward $2$ from values less than $a=2$ (left-hand side), the corresponding function values of $f(x) = \frac{x^2}{x-2}$ also increase and get closer to $4$. Similarly, as $x$ decreases toward 2 from values greater than $a=2$ (right-hand side) shown in the second table, the function values again approach $4$.

Overall, both sides of the table show values of $f(x) = \frac{x^2 - 4}{x - 2}$ getting closer to $4$ when $x$ approaches $2$. As the result, $ \displaystyle\lim_{{x \to 2}} \frac{x^2 - 4}{x - 2} = 4$


<h2>Python</h2>

In the cell below, there is code that:
1. Defines $f(x)$ as the function $\displaystyle f(x)=\frac{x^2-4}{x-2}$
2. Creates 2 arrays of x-values, one with x-values getting closer to $2$ *from the left side* and the other with x-values approaching 2 *from the right side*
3. Finds the corresponding y-values for the x-values created
4. Displays 2 tables representing the values of $y$ as $x$ approaches $2$ *from the left* AND a table that shows the values of $y$ as $x$ approaches 2 *from the right*

Please recall that in order for a limit to exist at a value of x, the y-values must approach the same value from the right AND the left.

```python
import numpy as np

# Define a function
def f(x):
    return (x**2 - 4) / (x - 2)  

#Create and array of x values that approach x=2 from above (right side of 2) and below (left side of) 2
x_values_left = np.linspace(1.8, 2, 5, endpoint=False)  # Values approaching 2 from the left, BUT NOT AT 2, since f(x) is UNdefined at x=2
x_values_right = np.linspace(2.001, 2.2, 5)            # Values approaching 2 from the right

#create and array by evaluating f(x) at the corresponding array of x values created

y_values_left = [f(x) for x in x_values_left]
y_values_right = [f(x) for x in x_values_right]

#display the table of values with x-values in one column and the corresponding y-values in the second volumn
#table of values from the left side of 2

print("Left-hand limit:")
print("x (left)\t f(x)")
for x, y in zip(x_values_left, y_values_left):
    print(f"{x:.3f}\t {y:.3f}")

#table of y-values as x approaches 2 from the right side
print("\nRight-hand limit:")
print("x (right)\t f(x)")
for x, y in zip(x_values_right, y_values_right):
    print(f"{x:.3f}\t {y:.3f}")
```

<h2> 3.2 One-Sided Limits </h2>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
  When we say that the limit of a function "does not exist" at a specific point, this description does not fully describe the function’s behavior at that point. Consider function $f(x) = \frac{|x-2|}{(x-2)}$.  When $x$ approachs to $2$, $f(x)$ does not approach a single value, so the limit $\lim_{x \to 2}f(x) = DNE$. Nevertheless, this statement by itself does not fully capture the function’s behavior near the $x-value$ of $2$. To provide a more accurate description, we introduce the idea of a **one-sided limit**. On the left side, as $x$ approachs $2$, $f(x)$ approaches $-1$. We can express this as:

$$\lim_{x \to 2^-}f(x) = -1$$

Similiarly, as $x$ approaches 2 from the right, $f(x)$ approaches 1:

$$\lim_{x \to 2^+}f(x) = 1$$
</div>

<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
<h2>Relating One-Sided and Two-Sided Limits</h2>
    
Let $f(x)$ be a function defined at all values in an open interval containing $a$, with the possible exception of $a$ itself, and let $L$ be a real number. Then:

<div style="text-align: center;">
  $\displaystyle\lim_{x \to a}f(x) = L$ if and only if $\displaystyle\lim_{x \to a^-}f(x) = L$ and $\displaystyle\lim_{x \to a^+}f(x) = L$
</div>

</br>  
You probably noticed that we had been informally using one-sided limits in the previous examples and questions when we looked at the numerical values of a function **from the right** or **from the left** of an $x$-value.  Now we have a more formal mathematical way of writing out what we are looking for.

<h2> 3.3 Infinite Limits </h2>


<div style=" color: black; padding: 16px; border-radius: 10px; font-family:'Segoe UI', sans-serif; box-shadow: 0 0 10px #2c5364;">
    An infinite limit describes the bahavior of a function when its values grow without bound (positive or negative) as $x$ approaches a certain point. 

We say: 

$$\lim_{x \to a} f(x) = \infty$$

If the values of $f(x)$ increase without bound as $x$ get closer to $a$ *from both sides*.

Similarly,

$$\lim_{x \to a} f(x) = -\infty$$
    
means the values of $f(x)$ decrease without bound as $x$ approaches $a$ *from both sides*.


</div>

  <h2 style="margin:0;">One-sided Infinite Limits</h2>
    
Sometimes, we care about what happens from only one side:

- Infinite limits from the left: if the values of $f(x)$ increase (or decrease) without bound as the values of $x$ (where $x<a$) approach the number $a$, then we say that limit as $x$ approaches $a$ from the left is positive (or negative) infinity and we write:

<div style="text-align: center;">
  $\displaystyle\lim_{x \to a^-}f(x) = +\infty$ or $\displaystyle\lim_{x \to a^-}f(x) = -\infty$
</div>

- Infinite limits from the right: if the values of $f(x)$ increase (or decrease) without bound as the values of $x$ (where $x>a$) approach the number $a$, then we say that limit as $x$ approaches $a$ from the right is positive (or negative) infinity and we write:

<div style="text-align: center;">
  $\displaystyle\lim_{x \to a^+}f(x) = +\infty$ or $\displaystyle\lim_{x \to a^+}f(x) = -\infty$
</div>


<h2>Example</h2>

Consider function $f(x) = \frac{1}{x-2}$. 

As $x$ approaches to $2$, the function values grow without bound. On the left-hand side $\displaystyle\lim_{x \to 2^-} f(x) = -\infty$ and on the right-hand side $\displaystyle\lim_{x \to 2^+} f(x) = \infty$.

This means the function has a vertical asymptote at $x=2$ (see the graph below).

  <h2 style="margin:0;">Infinite Limits from Positive Integers</h2>
    
If $n$ is a positive even integer, then

$$\lim_{x \to a} \frac{1}{(x-a)^n} = +\infty$$

If $n$ is a positive odd integer, then

$$\lim_{x \to a^+} \frac{1}{(x-a)^n} = +\infty$$

and

$$\lim_{x \to a^-} \frac{1}{(x-a)^n} = -\infty$$
</div>
